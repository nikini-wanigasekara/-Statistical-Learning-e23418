#Q3) Transformer Failiures

In a certain electricity distribution district, the number of emergency transformer failures reported in a single day is modeled as a discrete random variable $X$. Assume that $X$ takes values in the set $\{0,1,2,3\}$, with probabilities

$$
P(X=0)=0.50,\qquad P(X=1)=0.30,\qquad P(X=2)=0.15,\qquad P(X=3)=0.05.
$$

1. Construct a probability space $(\Omega,\mathfrak F,P)$ suitable for this model.
2. Define $X$ as a random variable on this space.
3. Determine the probability mass function of $X$.
4. Compute the expectation $E[X]$.
5. Compute the variance $\mathrm{Var}(X)$.
6. Find $P(X\ge 1)$ and $P(X\ge 2)$.
7. Briefly explain the practical meaning of the expectation and variance in this setting.



# Solution

## 1. Probability Space

Define the sample space as the set of possible daily failure counts:
$$\Omega = \{0,1,2,3\}.$$

Let $\mathfrak{F} = 2^{\Omega}$ be the power set of $\Omega$ (all $2^4 = 16$ subsets), which forms a $\sigma$-algebra on $\Omega$.

Define the probability measure $P: \mathfrak{F} \to [0,1]$ on the atoms by
$$P(\{0\})=0.50,\quad P(\{1\})=0.30,\quad P(\{2\})=0.15,\quad P(\{3\})=0.05,$$
and extend to all events by finite additivity. One verifies that $P(\Omega)=0.50+0.30+0.15+0.05=1$, so $(\Omega,\mathfrak{F},P)$ is a valid probability space. $\blacksquare$

---

## 2. Random Variable

Define $X:\Omega\to\mathbb{R}$ by the identity map:
$$X(\omega) = \omega, \qquad \omega\in\{0,1,2,3\}.$$
For every Borel set $B\subseteq\mathbb{R}$, the preimage $X^{-1}(B)=B\cap\Omega\in\mathfrak{F}$, so $X$ is $(\mathfrak{F},\mathcal{B}(\mathbb{R}))$-measurable and hence a valid random variable.

---

## 3. Probability Mass Function

The PMF $p:\mathbb{R}\to[0,1]$ of $X$ is
$$p(k)=P(X=k)=\begin{cases}0.50, & k=0,\\ 0.30, & k=1,\\ 0.15, & k=2,\\ 0.05, & k=3,\\ 0, & \text{otherwise.}\end{cases}$$

---

## 4. Expectation

By definition,
$$E[X]=\sum_{k=0}^{3}k\cdot P(X=k).$$

Computing term by term:

| $k$ | $P(X=k)$ | $k\cdot P(X=k)$ |
|-----|-----------|-----------------|
| 0   | 0.50      | 0.00            |
| 1   | 0.30      | 0.30            |
| 2   | 0.15      | 0.30            |
| 3   | 0.05      | 0.15            |

$$\boxed{E[X] = 0.00+0.30+0.30+0.15 = 0.75.}$$

---

## 5. Variance

First compute $E[X^2]$:
$$E[X^2]=\sum_{k=0}^{3}k^2\cdot P(X=k)=0(0.50)+1(0.30)+4(0.15)+9(0.05)=0+0.30+0.60+0.45=1.35.$$

Then by the computational formula $\mathrm{Var}(X)=E[X^2]-(E[X])^2$:
$$\boxed{\mathrm{Var}(X)=1.35-(0.75)^2=1.35-0.5625=0.7875.}$$

---

## 6. Tail Probabilities

**$P(X\ge 1)$:** Using the complement,
$$P(X\ge 1)=1-P(X=0)=1-0.50=\boxed{0.50.}$$

**$P(X\ge 2)$:** Similarly,
$$P(X\ge 2)=1-P(X=0)-P(X=1)=1-0.50-0.30=\boxed{0.20.}$$

Equivalently, $P(X\ge 2)=P(X=2)+P(X=3)=0.15+0.05=0.20$. $\checkmark$

---

## 7. Practical Interpretation

**Expectation** $E[X]=0.75$: On average, the district experiences $0.75$ emergency transformer failures per day. Over a long operational period of $n$ days, the total number of failures will be approximately $0.75n$ by the Law of Large Numbers — this figure directly informs staffing, spare-parts inventory, and maintenance scheduling.

**Variance** $\mathrm{Var}(X)=0.7875$: This measures the day-to-day variability around the average. The standard deviation $\sigma=\sqrt{0.7875}\approx 0.887$ is comparable in magnitude to $E[X]$, indicating **substantial relative uncertainty**: on any given day, actual failures may deviate considerably from the expected value. Engineers and planners should therefore maintain contingency capacity well beyond the average-demand level to ensure reliable service.

# Q4) Probability of Defective Items

In a manufacturing process, the proportion of defective items in a large batch produced on a given day is modeled by a continuous random variable $X$. Since $X$ represents a proportion, it takes values in the interval $[0,1]$. Suppose that $X$ has a Beta distribution with parameters $\alpha=2$ and $\beta=18$, so that its density is

$$
f(x)=
\begin{cases}
\dfrac{1}{B(2,18)}x^{1}(1-x)^{17}, & 0<x<1,\\\
0, & \text{otherwise},
\end{cases}
$$

where $B(2,18)$ is the Beta function.

Answer the following:

1. Construct a probability space $(\Omega,\mathfrak F,P)$ suitable for this model.

2. Define $X$ as a continuous random variable on this space.

3. Write down the probability density function of $X$.

4. Verify that $f$ is a valid probability density function.

5. Compute the expectation $E[X]$.

6. Compute the variance $\mathrm{Var}(X)$.

7. Find the probability that the defective proportion is at most $0.10$, that is, compute

$$
P(X\le 0.10).
$$

You may leave your answer in Beta integral form.

8. Find the probability that the defective proportion lies between $0.05$ and $0.15$, that is, compute

$$
P(0.05\le X\le 0.15).
$$

You may leave your answer in integral form.

9. Briefly explain why a Beta distribution is a reasonable model for this situation.

10. Interpret the expectation and variance in the context of production quality.


# Solution

## 1. Probability Space

Since $X$ represents a proportion, the natural sample space is the unit interval:
$$\Omega = [0,1].$$
Let $\mathfrak{F} = \mathcal{B}([0,1])$ be the Borel $\sigma$-algebra on $[0,1]$, generated by all sub-intervals of $[0,1]$.

Define the probability measure $P$ on $(\Omega, \mathfrak{F})$ by
$$P(A) = \int_A f(x)\,dx, \qquad A\in\mathfrak{F},$$
where $f$ is the Beta$(2,18)$ density given below. Since $f\ge 0$ and $\int_0^1 f(x)\,dx=1$ (verified in Part 4), this is a valid probability measure, and $(\Omega,\mathfrak{F},P)$ is a complete probability space. $\blacksquare$

---

## 2. Random Variable

Define $X:\Omega\to\mathbb{R}$ by the identity map:
$$X(\omega)=\omega, \qquad \omega\in[0,1].$$
For every Borel set $B\subseteq\mathbb{R}$, the preimage $X^{-1}(B)=B\cap[0,1]\in\mathcal{B}([0,1])=\mathfrak{F}$, so $X$ is $(\mathfrak{F},\mathcal{B}(\mathbb{R}))$-measurable and hence a valid continuous random variable.

---

## 3. Probability Density Function

Recalling the Beta function $B(\alpha,\beta)=\frac{(\alpha-1)!\,(\beta-1)!}{(\alpha+\beta-1)!}$ for positive integers, we have
$$B(2,18)=\frac{1!\cdot 17!}{19!}=\frac{1}{19\cdot 18}=\frac{1}{342},$$
so the PDF of $X\sim\mathrm{Beta}(2,18)$ is

$$\boxed{f(x)=\begin{cases}342\,x(1-x)^{17}, & 0<x<1,\\ 0, & \text{otherwise.}\end{cases}}$$

---

## 4. Validity of the Density

We must verify $f(x)\ge 0$ and $\int_{-\infty}^{\infty}f(x)\,dx=1$.

**Non-negativity.** For $x\in(0,1)$: $x>0$ and $(1-x)^{17}>0$, so $f(x)=342\,x(1-x)^{17}>0$. $\checkmark$

**Unit integral.** By definition of the Beta function and the relation to the Gamma function,
$$\int_0^1 x^{\alpha-1}(1-x)^{\beta-1}\,dx = B(\alpha,\beta)=\frac{\Gamma(\alpha)\Gamma(\beta)}{\Gamma(\alpha+\beta)}.$$
Therefore,
$$\int_{-\infty}^{\infty}f(x)\,dx = 342\int_0^1 x^{1}(1-x)^{17}\,dx = \frac{1}{B(2,18)}\cdot B(2,18)=1. \quad\checkmark$$

Hence $f$ is a valid probability density function. $\blacksquare$

---

## 5. Expectation

For $X\sim\mathrm{Beta}(\alpha,\beta)$, the general formula is $E[X]=\dfrac{\alpha}{\alpha+\beta}$. With $\alpha=2,\,\beta=18$:

$$\boxed{E[X]=\frac{2}{2+18}=\frac{2}{20}=0.10.}$$

*Direct verification:*
$$E[X]=\frac{1}{B(2,18)}\int_0^1 x^2(1-x)^{17}\,dx = \frac{B(3,18)}{B(2,18)}=\frac{\Gamma(3)\Gamma(18)/\Gamma(21)}{\Gamma(2)\Gamma(18)/\Gamma(20)}=\frac{2!\cdot 19!}{1!\cdot 20!}=\frac{2}{20}=0.10.\;\checkmark$$

---

## 6. Variance

For $X\sim\mathrm{Beta}(\alpha,\beta)$, the general formula is $\mathrm{Var}(X)=\dfrac{\alpha\beta}{(\alpha+\beta)^2(\alpha+\beta+1)}$. With $\alpha=2,\,\beta=18$:

$$\mathrm{Var}(X)=\frac{2\cdot 18}{(20)^2(21)}=\frac{36}{400\cdot 21}=\frac{36}{8400}=\frac{3}{700}.$$

$$\boxed{\mathrm{Var}(X)=\frac{3}{700}\approx 0.00429.}$$

---

## 7. Probability $P(X\le 0.10)$

$$P(X\le 0.10)=\int_0^{0.10}342\,x(1-x)^{17}\,dx.$$

This equals the **regularized incomplete Beta function** $I_{0.10}(2,18)$:

$$\boxed{P(X\le 0.10)=\frac{B_{0.10}(2,18)}{B(2,18)}=I_{0.10}(2,18)\approx 0.2639,}$$

where $B_t(\alpha,\beta)=\int_0^t x^{\alpha-1}(1-x)^{\beta-1}\,dx$ is the incomplete Beta function.

> **Remark.** The numerical value $I_{0.10}(2,18)\approx 0.2639$ is obtained via standard statistical software or tables; the integral form is the exact closed-form answer.

---

## 8. Probability $P(0.05\le X\le 0.15)$

$$\boxed{P(0.05\le X\le 0.15)=\int_{0.05}^{0.15}342\,x(1-x)^{17}\,dx = I_{0.15}(2,18)-I_{0.05}(2,18)\approx 0.3222.}$$

where the numerical approximation follows from standard Beta CDF tables or software.

---

## 9. Why the Beta Distribution is a Reasonable Model

The Beta distribution is the canonical choice for modeling a random proportion for the following reasons:

- **Support.** The Beta distribution has support exactly on $(0,1)$, matching the constraint that a proportion must lie in $[0,1]$, unlike the Normal or Gamma distributions which require truncation.

- **Flexibility.** Through the two shape parameters $\alpha$ and $\beta$, the Beta family encompasses a wide variety of shapes — skewed left, skewed right, symmetric, U-shaped, or uniform — allowing it to fit empirical defect-rate data closely.

- **Conjugate prior.** In Bayesian quality control, the Beta distribution is the conjugate prior for the binomial likelihood, so posterior updates after observing defective counts remain analytically tractable.

- **Parameter interpretation.** The choice $\alpha=2,\,\beta=18$ encodes the prior belief that defectives are rare ($\alpha\ll\beta$), producing a right-skewed density concentrated near $x=0$, consistent with a well-functioning manufacturing process.

---

## 10. Interpretation in the Context of Production Quality

**Expectation** $E[X]=0.10$: On a typical production day, $10\%$ of items in the batch are expected to be defective. This is the long-run average defect rate across many production days, providing the central planning benchmark for quality targets and acceptance sampling schemes.

**Variance** $\mathrm{Var}(X)=\frac{3}{700}\approx 0.0043$, giving a standard deviation of $\sigma\approx 0.065$ (i.e., $6.5$ percentage points). This quantifies day-to-day variability in the defect rate. Despite the average being $10\%$, on any given day the proportion of defectives may plausibly range from near $0\%$ to above $20\%$. A quality engineer should design control limits and inspection procedures that account for this spread, rather than treating the expected rate as a fixed constant.

# Q5) Expected Project Completion Time

A construction company is estimating the time required to install a backup generator at a hospital. Let $X$ denote the number of hours needed to complete the installation. Based on expert judgment, the shortest possible time is $4$ hours, the longest possible time is $10$ hours, and the most likely completion time is $6$ hours. Assume that $X$ follows a Triangular distribution with parameters $a=4$, $b=10$, and $c=6$.

Answer the following:

1. Specify the support of $X$.

2. Write down the probability density function $f_X(x)$.

3. Verify that $f_X$ integrates to $1$ over its support.

4. Compute the expectation $E[X]$.

5. Compute the variance $\mathrm{Var}(X)$.

6. Interpret the values of $a$, $b$, and $c$ in the context of this problem.

7. Explain briefly why a Triangular distribution is a reasonable model in this setting.

8. Compute the probability that the installation takes at most $5$ hours, that is, compute

$$
P(X \leq 5).
$$

9. Compute the probability that the installation takes between $5$ and $8$ hours, that is, compute

$$
P(5 \leq X \leq 8).
$$

10. Briefly explain how this model could be used in project planning or sensitivity analysis.


# Solution

## 1. Support of $X$

Since the installation time is bounded below by the minimum $a=4$ and above by the maximum $b=10$, the support of $X$ is the closed interval

$$\operatorname{supp}(X)=[a,b]=[4,10].$$

---

## 2. Probability Density Function

For a Triangular distribution with minimum $a$, maximum $b$, and mode $c$, the PDF is

$$f_X(x)=\begin{cases}\dfrac{2(x-a)}{(b-a)(c-a)}, & a\le x\le c,\\[10pt] \dfrac{2(b-x)}{(b-a)(b-c)}, & c < x\le b,\\[6pt] 0, & \text{otherwise.}\end{cases}$$

Substituting $a=4$, $b=10$, $c=6$:

$$\boxed{f_X(x)=\begin{cases}\dfrac{2(x-4)}{(6)(2)}=\dfrac{x-4}{6}, & 4\le x\le 6,\\[10pt] \dfrac{2(10-x)}{(6)(4)}=\dfrac{10-x}{12}, & 6 < x\le 10,\\[6pt] 0, & \text{otherwise.}\end{cases}}$$

---

## 3. Verification that $f_X$ Integrates to 1

Split the integral at the mode $c=6$:

$$\int_{-\infty}^{\infty}f_X(x)\,dx = \underbrace{\int_4^6\frac{x-4}{6}\,dx}_{I_1}+\underbrace{\int_6^{10}\frac{10-x}{12}\,dx}_{I_2}.$$

**Computing $I_1$:** Let $u=x-4$, so $du=dx$ and the limits become $0$ to $2$:
$$I_1=\frac{1}{6}\int_0^2 u\,du=\frac{1}{6}\cdot\frac{u^2}{2}\Bigg|_0^2=\frac{1}{6}\cdot 2=\frac{1}{3}.$$

**Computing $I_2$:** Let $v=10-x$, so $dv=-dx$ and the limits become $4$ to $0$:
$$I_2=\frac{1}{12}\int_0^4 v\,dv=\frac{1}{12}\cdot\frac{v^2}{2}\Bigg|_0^4=\frac{1}{12}\cdot 8=\frac{2}{3}.$$

**Total:**
$$I_1+I_2=\frac{1}{3}+\frac{2}{3}=1.\quad\checkmark\qquad\blacksquare$$

---

## 4. Expectation

For the Triangular distribution, the general formula is

$$E[X]=\frac{a+b+c}{3}.$$

Substituting:

$$\boxed{E[X]=\frac{4+10+6}{3}=\frac{20}{3}\approx 6.67\text{ hours.}}$$

*Direct verification:*
$$E[X]=\int_4^6\frac{x(x-4)}{6}\,dx+\int_6^{10}\frac{x(10-x)}{12}\,dx.$$

**First integral:**
$$\int_4^6\frac{x^2-4x}{6}\,dx=\frac{1}{6}\left[\frac{x^3}{3}-2x^2\right]_4^6=\frac{1}{6}\left[\left(72-72\right)-\left(\frac{64}{3}-32\right)\right]=\frac{1}{6}\cdot\frac{32}{3}=\frac{16}{9}.$$

**Second integral:**
$$\int_6^{10}\frac{10x-x^2}{12}\,dx=\frac{1}{12}\left[5x^2-\frac{x^3}{3}\right]_6^{10}=\frac{1}{12}\left[\left(500-\frac{1000}{3}\right)-\left(180-72\right)\right]=\frac{1}{12}\cdot\frac{284}{3}=\frac{71}{9}.$$

$$E[X]=\frac{16}{9}+\frac{71}{9}=\frac{87}{9}=\frac{20}{3}.\quad\checkmark$$

---

## 5. Variance

For the Triangular distribution, the general formula is

$$\mathrm{Var}(X)=\frac{a^2+b^2+c^2-ab-ac-bc}{18}.$$

Substituting $a=4,\,b=10,\,c=6$:

$$\mathrm{Var}(X)=\frac{16+100+36-40-24-60}{18}=\frac{28}{18}=\frac{14}{9}.$$

$$\boxed{\mathrm{Var}(X)=\frac{14}{9}\approx 1.556\text{ hours}^2,\qquad \sigma_X=\sqrt{\frac{14}{9}}\approx 1.247\text{ hours.}}$$

---

## 6. Interpretation of the Parameters

| Parameter | Value | Meaning in context |
|-----------|-------|--------------------|
| $a=4$ | 4 hours | **Minimum:** the absolute best-case scenario — ideal conditions, no complications whatsoever. Installation cannot be completed any faster. |
| $b=10$ | 10 hours | **Maximum:** the worst-case scenario — significant delays, equipment complications, or logistical issues. Installation will not exceed this time. |
| $c=6$ | 6 hours | **Mode:** the single most likely completion time under normal working conditions, as judged by experienced engineers. |

The asymmetry $c-a=2 < b-c=4$ indicates that the distribution is **right-skewed**: the installation is more likely to finish close to the optimistic end, but the upper tail is longer — there is greater scope for delay than for beating the most likely time.

---

## 7. Why the Triangular Distribution is a Reasonable Model

The Triangular distribution is particularly well-suited to this setting for the following reasons:

- **Expert elicitation.** When historical data are scarce or unavailable — as is common for one-off construction projects — engineers can readily estimate three intuitive quantities: the minimum, maximum, and most likely duration. The Triangular distribution is fully specified by exactly these three values, making it directly applicable.

- **Bounded support.** Physical installation times cannot be negative or unbounded; the $[a,b]$ support enforces hard constraints consistent with engineering reality.

- **Asymmetry.** Unlike the symmetric Uniform or Normal distributions, the Triangular distribution naturally accommodates right-skewed completion times, reflecting the practical observation that delays are more common and more severe than early completions.

- **Simplicity and transparency.** The piecewise-linear PDF is easy to communicate to non-statistician stakeholders such as project managers, site engineers, and hospital administrators.

---

## 8. Probability $P(X\le 5)$

Since $5\in[4,6]$, we use the left branch of $f_X$:

$$P(X\le 5)=\int_4^5\frac{x-4}{6}\,dx=\frac{1}{6}\cdot\frac{(x-4)^2}{2}\Bigg|_4^5=\frac{1}{12}(5-4)^2=\frac{1}{12}.$$

$$\boxed{P(X\le 5)=\frac{1}{12}\approx 0.0833.}$$

This means there is roughly an **8.3% chance** the installation is completed within the first hour beyond the minimum — a relatively unlikely outcome.

---

## 9. Probability $P(5\le X\le 8)$

Since $5<c=6<8$, we split at the mode:

$$P(5\le X\le 8)=\int_5^6\frac{x-4}{6}\,dx+\int_6^8\frac{10-x}{12}\,dx.$$

**Left piece** ($5$ to $6$, left branch):
$$\int_5^6\frac{x-4}{6}\,dx=\frac{1}{12}(x-4)^2\Big|_5^6=\frac{1}{12}(4-1)=\frac{3}{12}=\frac{1}{4}.$$

**Right piece** ($6$ to $8$, right branch):
$$\int_6^8\frac{10-x}{12}\,dx=\frac{-1}{24}(10-x)^2\Big|_6^8=\frac{-1}{24}(4-16)=\frac{12}{24}=\frac{1}{2}.$$

**Total:**
$$\boxed{P(5\le X\le 8)=\frac{1}{4}+\frac{1}{2}=\frac{3}{4}=0.75.}$$

There is a **75% probability** that the installation is completed within the 5–8 hour window — the bulk of the distribution's probability mass.

---

## 10. Application to Project Planning and Sensitivity Analysis

**Project planning.** The model directly informs scheduling decisions for the hospital:

- The expected completion time $E[X]=\tfrac{20}{3}\approx 6.67$ hours sets the baseline for allocating crew hours, booking equipment, and scheduling hospital downtime.
- The standard deviation $\sigma\approx 1.25$ hours quantifies scheduling risk; a project manager might add a **time buffer** of one or two standard deviations — scheduling an 8–9 hour window — to achieve high confidence that the installation finishes on time.
- Computed probabilities such as $P(X\le 5)\approx 8.3\%$ and $P(X\le 8)$ allow the team to set **contractual deadlines** at a desired confidence level (e.g., the 90th percentile completion time).

**Sensitivity analysis.** The Triangular distribution integrates naturally into **Monte Carlo simulation** for larger project networks (e.g., PERT/CPM):

- By sampling $X\sim\mathrm{Triangular}(4,10,6)$ repeatedly, planners can simulate the full distribution of total project completion time when generator installation is one task among many interdependent tasks.
- Analysts can test how the overall project schedule changes if expert estimates are revised — for instance, if supply-chain issues shift the mode from $c=6$ to $c=8$ — thereby identifying which tasks carry the most schedule risk and deserve the tightest management attention.

#Q6) Expected Project Completion Time Using PERT

A project manager is estimating the time required to complete a critical software integration task. Let $X$ denote the number of days required to complete the task. Based on expert judgment, the shortest possible time is $a=8$ days, the longest possible time is $b=20$ days, and the most likely completion time is $c=11$ days. Assume that $X$ follows a PERT distribution with parameters $(a,b,c)$.

The PERT distribution is defined as a Beta distribution scaled to the interval $[a,b]$, with internal shape parameters

$$
\alpha = 1 + 4\left( \frac{c-a}{b-a} \right)
$$

and

$$
\beta = 1 + 4\left( \frac{b-c}{b-a} \right).
$$

Its probability density function is

$$
f(x)=\frac{(x-a)^{\alpha-1}(b-x)^{\beta-1}}{B(\alpha,\beta)(b-a)^{\alpha+\beta-1}}, \qquad a \leq x \leq b,
$$

where

$$
B(\alpha,\beta)=\int_0^1 t^{\alpha-1}(1-t)^{\beta-1},dt.
$$

Answer the following:

1. Compute the values of the shape parameters $\alpha$ and $\beta$.

2. Specify the support of $X$.

3. Write down the probability density function $f_X(x)$ for this model.

4. State the probability space $(\Omega,\mathfrak F,P)$ suitable for this continuous model.

5. Define $X$ as a random variable on this space.

6. Compute the expectation $E[X]$ using the PERT formula

$$
E[X]=\frac{a+4c+b}{6}.
$$

7. Compute the variance $\mathrm{Var}(X)$ using the approximation

$$
\mathrm{Var}(X)=\frac{(E[X]-a)(b-E[X])}{7}.
$$

8. Compute the approximate standard deviation $\sigma$.

9. Briefly explain why a PERT distribution may be more appropriate than a Triangular distribution in this setting byt comparing the two results.

10. Interpret the practical meaning of the values $a$, $b$, $c$, $E[X]$, and $\sigma$ in the context of project planning.


# Solution

## 1. Shape Parameters $\alpha$ and $\beta$

With $a=8$, $b=20$, $c=11$, the range is $b-a=12$. Substituting into the PERT formulas:

$$\alpha = 1+4\left(\frac{c-a}{b-a}\right)=1+4\left(\frac{11-8}{12}\right)=1+4\cdot\frac{3}{12}=1+1=2,$$

$$\beta = 1+4\left(\frac{b-c}{b-a}\right)=1+4\left(\frac{20-11}{12}\right)=1+4\cdot\frac{9}{12}=1+3=4.$$

$$\boxed{\alpha=2,\qquad \beta=4.}$$

---

## 2. Support of $X$

Since the PERT distribution is a Beta distribution scaled to $[a,b]$:
$$\operatorname{supp}(X)=[a,b]=[8,20].$$

---

## 3. Probability Density Function

First evaluate the necessary constants. For positive-integer parameters, $B(\alpha,\beta)=\frac{(\alpha-1)!\,(\beta-1)!}{(\alpha+\beta-1)!}$:
$$B(2,4)=\frac{1!\cdot 3!}{5!}=\frac{6}{120}=\frac{1}{20},$$
and $(b-a)^{\alpha+\beta-1}=12^{2+4-1}=12^5=248832.$

Substituting into the general PERT density:

$$\boxed{f_X(x)=\frac{(x-8)^{1}(20-x)^{3}}{\dfrac{1}{20}\cdot 248832}=\frac{20(x-8)(20-x)^3}{248832}=\frac{5(x-8)(20-x)^3}{62208},\quad 8\le x\le 20.}$$

---

## 4. Probability Space

Define the sample space, $\sigma$-algebra, and measure as follows:
$$\Omega=[8,20], \qquad \mathfrak{F}=\mathcal{B}([8,20]), \qquad P(A)=\int_A f_X(x)\,dx \quad \text{for } A\in\mathfrak{F},$$
where $\mathcal{B}([8,20])$ is the Borel $\sigma$-algebra on $[8,20]$. Since $f_X\ge 0$ and $\int_8^{20}f_X(x)\,dx=1$ (by construction of the Beta function), $P$ is a valid probability measure and $(\Omega,\mathfrak{F},P)$ is a proper probability space. $\blacksquare$

---

## 5. Random Variable

Define $X:\Omega\to\mathbb{R}$ by the identity map:
$$X(\omega)=\omega,\qquad \omega\in[8,20].$$
For every Borel set $B\subseteq\mathbb{R}$, $X^{-1}(B)=B\cap[8,20]\in\mathcal{B}([8,20])=\mathfrak{F}$, so $X$ is $(\mathfrak{F},\mathcal{B}(\mathbb{R}))$-measurable and hence a valid continuous random variable.

---

## 6. Expectation

Applying the PERT formula directly:
$$E[X]=\frac{a+4c+b}{6}=\frac{8+4(11)+20}{6}=\frac{8+44+20}{6}=\frac{72}{6}.$$

$$\boxed{E[X]=12\text{ days.}}$$

> **Remark.** The PERT formula weights the mode fourfold relative to each endpoint, reflecting the practical judgment that expert estimates of the most likely value are more reliable than extreme bounds.

---

## 7. Variance

Substituting $E[X]=12$, $a=8$, $b=20$ into the PERT approximation:

$$\mathrm{Var}(X)=\frac{(E[X]-a)(b-E[X])}{7}=\frac{(12-8)(20-12)}{7}=\frac{4\cdot 8}{7}=\frac{32}{7}.$$

$$\boxed{\mathrm{Var}(X)=\frac{32}{7}\approx 4.571\text{ days}^2.}$$

---

## 8. Standard Deviation

$$\boxed{\sigma=\sqrt{\mathrm{Var}(X)}=\sqrt{\frac{32}{7}}=\frac{4\sqrt{2}}{\sqrt{7}}=4\sqrt{\frac{2}{7}}\approx 2.138\text{ days.}}$$

---

## 9. PERT vs. Triangular Distribution

To make the comparison concrete, we compute what the Triangular distribution with the same parameters $(a,b,c)=(8,20,11)$ would give.

| Quantity | Triangular$(8,20,11)$ | PERT$(8,20,11)$ |
|---|---|---|
| $E[X]$ | $\dfrac{a+b+c}{3}=\dfrac{39}{3}=13$ days | $\dfrac{a+4c+b}{6}=12$ days |
| $\mathrm{Var}(X)$ | $\dfrac{a^2+b^2+c^2-ab-ac-bc}{18}=\dfrac{28}{3}\approx 9.33$ days$^2$ | $\dfrac{32}{7}\approx 4.57$ days$^2$ |
| $\sigma$ | $\approx 3.055$ days | $\approx 2.138$ days |

The two distributions differ in two important structural ways:

**Expectation.** The Triangular mean weights all three parameters equally, giving $E[X]=13$ days. The PERT mean weights the mode $c$ fourfold, pulling the mean toward the most expert-judged value and yielding $E[X]=12$ days — closer to $c=11$. This is more appropriate when expert judgment about the mode is reliable.

**Variance and tail behaviour.** The Triangular density has a sharp triangular peak at $c$ and drops linearly to zero at the endpoints, producing relatively fat tails and a larger variance ($\approx 9.33$ days$^2$). The PERT density, being a smooth Beta curve, tapers more gently and concentrates more probability mass near the mode, giving a substantially smaller variance ($\approx 4.57$ days$^2$) and standard deviation. In practice, task durations rarely jump abruptly to their extremes; the smooth PERT curve is a more faithful model of this behaviour.

**Conclusion.** The PERT distribution is preferable here because it honours expert mode estimates more strongly, produces a smoother and more realistic density shape, and yields tighter, more credible uncertainty estimates for project scheduling.

## 10. Practical Interpretation in Project Planning

| Quantity | Value | Practical meaning |
|---|---|---|
| $a=8$ | 8 days | **Optimistic bound.** The absolute minimum — achievable only under ideal conditions with no obstacles. Useful for setting best-case contractual milestones. |
| $b=20$ | 20 days | **Pessimistic bound.** The worst-case ceiling — accounting for foreseeable delays, integration failures, and rework cycles. Useful for contingency budgeting and deadline risk assessment. |
| $c=11$ | 11 days | **Most likely duration.** The mode under normal working conditions as judged by the expert. This is the value communicated in day-to-day planning conversations. |
| $E[X]=12$ | 12 days | **Expected duration.** The probability-weighted average completion time. Project schedules, resource allocation, and budget forecasts should be anchored to this figure rather than the mode, since it accounts for the right-skewed risk of delay. |
| $\sigma\approx 2.14$ | $\approx 2.14$ days | **Scheduling uncertainty.** One standard deviation around the mean. A risk-aware project manager would add a buffer of $1$–$2\sigma$ to the baseline schedule — targeting completion by day $12+2(2.14)\approx 16.3$ — to achieve high confidence of on-time delivery. |

**Monte Carlo integration.** In a full project network with multiple interdependent tasks, each task duration is modelled as an independent PERT random variable. The total project duration is the sum (or critical-path maximum) of these variables. By simulating thousands of realisations, the project manager obtains the full distribution of project completion time, identifies the probability of meeting a given deadline, and pinpoints which tasks carry the greatest schedule risk — enabling targeted risk mitigation and resource reallocation.

#Q7) If the PERT Expectation is to Match Our Intuition

A project manager models the duration $X$ of a task by a PERT distribution with parameters $a$, $b$, and $c$, where $a$ is the minimum time, $b$ is the maximum time, and $c$ is the most likely time.

Suppose the minimum time is $a=5$ days and the most likely time is $c=11$ days. If the expected time computed from the PERT model is also to be the same as the most likely time:

1. Determine the value of the maximum time $b$.

2. What happens to the variance?

# Solution

## 1. Determining $b$

The PERT expectation formula gives
$$E[X]=\frac{a+4c+b}{6}.$$

We impose the condition $E[X]=c$, so
$$c = \frac{a+4c+b}{6}.$$

Multiplying both sides by $6$:
$$6c = a+4c+b.$$

Solving for $b$:
$$b = 6c-4c-a = 2c-a.$$

Substituting $a=5$ and $c=11$:
$$\boxed{b = 2(11)-5 = 17\text{ days.}}$$

---

## 2. Behaviour of the Variance

Substituting $E[X]=c$, $a=5$, $b=17$ into the PERT variance formula:

$$\mathrm{Var}(X)=\frac{(E[X]-a)(b-E[X])}{7}=\frac{(c-a)(b-c)}{7}.$$

Now observe that under the condition $b=2c-a$:
$$b-c = (2c-a)-c = c-a.$$

Therefore:
$$\boxed{\mathrm{Var}(X)=\frac{(c-a)^2}{7}=\frac{(11-5)^2}{7}=\frac{36}{7}\approx 5.14\text{ days}^2.}$$

**Structural observation.** The condition $E[X]=c$ forces $c-a=b-c$, meaning the mode $c$ lies exactly at the **midpoint** of $[a,b]$. The distribution is therefore **symmetric** about $c$, and the variance depends only on the half-range $(c-a)$: wider expert bounds inflate uncertainty proportionally, while a tighter range reduces it, regardless of where $c$ sits within $[a,b]$.

# Q1) The Probability of Shooting at a Target

A point is chosen at random from a disk of radius $10$. Let $A$ be the event that the point lies within $1$ unit of the boundary.

1. Model the experiment assuming the point is uniformly distributed with respect to area, and compute $P(A)$.
2. Model the experiment assuming the distance to the point from the center is chosen uniformly from $[0,10]$, and the direction is chosen independently and uniformly from $[0,2\pi)$ and compute $P(A)$.
3. Explain why the two answers differ.




## 1. Uniform Distribution with Respect to Area
In this model, we assume the probability of the point landing in a region is proportional to that region's **area**.

* **Total Area ($S$):** The area of a disk with radius $R = 10$.
    $$Area(S) = \pi \times 10^2 = 100\pi$$
* **Event $A$:** The point lies within 1 unit of the boundary. This means the distance $r$ from the center is between 9 and 10 ($9 \le r \le 10$).
* **Area of $A$:** This is the area of the total disk minus the area of the inner disk (radius 9).
    $$Area(A) = \pi(10^2) - \pi(9^2) = 100\pi - 81\pi = 19\pi$$
* **Probability $P(A)$:**
    $$P(A) = \frac{Area(A)}{Area(S)} = \frac{19\pi}{100\pi} = 0.19$$

---

## 2. Uniform Distribution with Respect to Radius
In this model, the distance $R$ is a random variable $R \sim U(0, 10)$, and the angle $\theta$ is independent. The geometry of the circle (area) does not weight the probability.

* **Total Sample Space:** The possible values for the radius $r$ are in the interval $[0, 10]$, which has a total length of **10**.
* **Event $A$:** The distance $r$ is within 1 unit of the boundary, meaning $r \in [9, 10]$. This interval has a length of **1**.
* **Probability $P(A)$:**
    $$P(A) = \frac{\text{Length of favorable interval}}{\text{Total length of radius interval}} = \frac{10 - 9}{10} = 0.1$$

---

## 3. Why the Answers Differ
The difference lies in how "uniformity" is defined in each coordinate system.

1.  **Area-Uniformity:** Recognizes that as you move further from the center, each "radial increment" covers more physical space. Because there is significantly more surface area near the edge of a disk than near the center, a point chosen truly at random by location is more likely to be near the boundary.
2.  **Radius-Uniformity:** Ignores the physical area. It treats the center of the circle as a single point that is just as "likely" to be chosen as the entire outer ring. This effectively "bunches" or concentrates the probability density toward the center of the disk when mapped onto a 2D plane.



This is a classic example of **Bertrand's Paradox**, which demonstrates that "random" is not well-defined until you specify the mechanism or distribution used to select the random variables.

# Q2) Bertrand's Paradox

A chord of a circle is chosen “at random.” What is the probability that its length exceeds the side length of the inscribed equilateral triangle?

Show that different mathematically natural models of “random chord” lead to different answers.



### **1. The Random Endpoints Method**
Choose two points on the circumference independently and uniformly. Rotate the circle so the first point is at a vertex of an inscribed equilateral triangle. The chord exceeds the triangle's side length only if the second point falls on the arc between the other two vertices.
* **Calculation:** The arc length is $1/3$ of the circumference.
* **Probability:** $P = 1/3$



### **2. The Random Radial Distance Method**
Choose a radius and a point on that radius uniformly. Construct a chord perpendicular to the radius at that point. The chord exceeds the side of the equilateral triangle if the chosen point is closer to the center than the midpoint of the radius (where the triangle side would intersect it).
* **Calculation:** The side of the inscribed triangle is at distance $R/2$ from the center. We need the distance $d < R/2$ within the range $[0, R]$.
* **Probability:** $P = 1/2$



### **3. The Random Midpoint Method**
Choose a point anywhere inside the circle uniformly as the chord's midpoint. For the chord length to exceed the triangle's side, the midpoint must fall within a smaller concentric circle that is tangent to the triangle's sides.
* **Calculation:** The radius of this inner circle is $r = R/2$. The probability is the ratio of the areas:
$$\frac{\text{Area}_{\text{inner}}}{\text{Area}_{\text{outer}}} = \frac{\pi (R/2)^2}{\pi R^2} = \frac{1}{4}$$
* **Probability:** $P = 1/4$

#Q8) Probability of Being a Girl?

A family has two children. Assume each child is equally likely to be a boy or a girl, independently of the other.

What is the probability that both children are girls, given that **at least one** of them is a girl?

The probability is **1/3** (approximately **0.333**).

### Explanation
When you are told that at least one child is a girl, you are eliminating the possibility of having two boys (**BB**).

Initially, there are four equally likely possibilities for a two-child family:
1.  **BG** (Boy, then Girl)
2.  **GB** (Girl, then Boy)
3.  **GG** (Girl, then Girl)
4.  **BB** (Boy, then Boy) — *Eliminated by the condition*



Since we know there is "at least one girl," only the first three scenarios are possible. Out of those **3 remaining possibilities**, only **1** scenario (**GG**) results in both children being girls.

Thus, the probability is:
$$P(\text{Both Girls} \mid \text{At least one Girl}) = \frac{1}{3}$$

# Q9) Information Available in a Discrete Random Variable

A factory inspects a product and records its quality using two different coding systems.

Each product falls into exactly one of the following four categories:

* $G_1$: good quality, produced on Machine 1
* $G_2$: good quality, produced on Machine 2
* $D_1$: defective, produced on Machine 1
* $D_2$: defective, produced on Machine 2

Assume the sample space is

$$
\Omega=\{G_1,G_2,D_1,D_2\},
$$

with probabilities

$$
P(\{G_1\})=0.50,\qquad P(\{G_2\})=0.20,\qquad P(\{D_1\})=0.10,\qquad P(\{D_2\})=0.20.
$$

Define two discrete random variables on the same probability space:

* $X$ records only whether the item is good or defective:
  $$
  X(G_1)=0,\quad X(G_2)=0,\quad X(D_1)=1,\quad X(D_2)=1.
  $$

* $Y$ records both quality and machine:
  $$
  Y(G_1)=1,\quad Y(G_2)=2,\quad Y(D_1)=3,\quad Y(D_2)=4.
  $$

Answer the following.

1. Write down the probability space $(\Omega,\mathfrak F,P)$, where $\mathfrak F=\mathfrak P(\Omega)$.

2. Compute the $\sigma$-algebra generated by $X$, namely $\sigma(X)$.

3. Compute the $\sigma$-algebra generated by $Y$, namely $\sigma(Y)$.

4. Show that $\sigma(X)\subseteq \sigma(Y)$, and explain why $Y$ contains more information than $X$.

5. Find the marginal distribution of $X$.

6. Find the marginal distribution of $Y$.

7. Compute the joint probabilities
   $$
   P(X=0,Y=1),\quad P(X=0,Y=2),\quad P(X=1,Y=3),\quad P(X=1,Y=4).
   $$

8. Compute the conditional probabilities
   $$
   P(X=1\mid Y=3),\qquad P(X=1\mid Y=4),\qquad P(Y=3\mid X=1),\qquad P(Y=4\mid X=1).
   $$

9. Interpret the difference between conditioning on $X$ and conditioning on $Y$.

10. Explain, in words, what information is available in $\sigma(X)$ but not in $\sigma(Y)$, and what information is available in $\sigma(Y)$ but not in $\sigma(X)$.

11. Give a Python based visualization of the model.

12. Find the marginal, joint, and conditional Shannon Entropy of $X$ and $Y$.

### 1. The Probability Space $(\Omega, \mathcal{F}, P)$
* **Sample Space:** $\Omega = \{G_1, G_2, D_1, D_2\}$
* **$\sigma$-algebra:** $\mathcal{F} = \mathcal{P}(\Omega)$ (the power set, containing all $2^4 = 16$ possible subsets).
* **Probability Measure $P$:** Defined by the atoms:
    * $P(\{G_1\}) = 0.50$
    * $P(\{G_2\}) = 0.20$
    * $P(\{D_1\}) = 0.10$
    * $P(\{D_2\}) = 0.20$

### 2. $\sigma$-algebra generated by $X$, $\sigma(X)$
$X$ partitions $\Omega$ into $\{G_1, G_2\}$ (where $X=0$) and $\{D_1, D_2\}$ (where $X=1$).
$$\sigma(X) = \{ \emptyset, \{G_1, G_2\}, \{D_1, D_2\}, \Omega \}$$

### 3. $\sigma$-algebra generated by $Y$, $\sigma(Y)$
$Y$ assigns a unique value to every outcome in $\Omega$. Therefore, it generates the finest possible partition.
$$\sigma(Y) = \mathcal{P}(\Omega) \text{ (All 16 subsets of } \Omega)$$

### 4. Relationship and Information
* **Proof:** Since every set in $\sigma(X)$ (e.g., $\{G_1, G_2\}$) can be formed by the union of sets in $\sigma(Y)$ (e.g., $\{G_1\} \cup \{G_2\}$), it follows that $\sigma(X) \subseteq \sigma(Y)$.
* **Reasoning:** $Y$ identifies both the quality and the specific machine, whereas $X$ only identifies quality. Knowing the value of $Y$ tells you the value of $X$ with certainty, but knowing $X$ does not tell you the value of $Y$.

### 5. Marginal Distribution of $X$
* $P(X=0) = P(\{G_1, G_2\}) = 0.50 + 0.20 = 0.70$
* $P(X=1) = P(\{D_1, D_2\}) = 0.10 + 0.20 = 0.30$

### 6. Marginal Distribution of $Y$
* $P(Y=1) = 0.50$
* $P(Y=2) = 0.20$
* $P(Y=3) = 0.10$
* $P(Y=4) = 0.20$

### 7. Joint Probabilities
* $P(X=0, Y=1) = P(\{G_1\}) = 0.50$
* $P(X=0, Y=2) = P(\{G_2\}) = 0.20$
* $P(X=1, Y=3) = P(\{D_1\}) = 0.10$
* $P(X=1, Y=4) = P(\{D_2\}) = 0.20$
*(Note: All other combinations like $P(X=0, Y=3)$ are 0).*

### 8. Conditional Probabilities
* $P(X=1 \mid Y=3) = \frac{P(X=1, Y=3)}{P(Y=3)} = \frac{0.10}{0.10} = 1$
* $P(X=1 \mid Y=4) = \frac{P(X=1, Y=4)}{P(Y=4)} = \frac{0.20}{0.20} = 1$
* $P(Y=3 \mid X=1) = \frac{P(X=1, Y=3)}{P(X=1)} = \frac{0.10}{0.30} = 1/3 \approx 0.333$
* $P(Y=4 \mid X=1) = \frac{P(X=1, Y=4)}{P(X=1)} = \frac{0.20}{0.30} = 2/3 \approx 0.667$

### 9. Interpretation of Conditioning
Conditioning on $Y$ provides **full certainty** about the state of the product (machine and quality). Conditioning on $X$ only narrows the state down to a subset (e.g., if it's defective, it could still be from Machine 1 or Machine 2), leaving some residual uncertainty.

### 10. Information Availability
* **In $\sigma(X)$ but not $\sigma(Y)$:** Nothing. Since $\sigma(X) \subseteq \sigma(Y)$, all information in $X$ is already contained within $Y$.
* **In $\sigma(Y)$ but not $\sigma(X)$:** The specific **Machine ID**. $Y$ distinguishes between $G_1$ and $G_2$, while $X$ groups them both as "0".

### 11. Visualization Description
A visualization would typically show a bar chart of the joint distribution $P(X,Y)$. Since $X$ is functionally dependent on $Y$, the heatmap or 3D bar chart would show non-zero values only at coordinates $(0,1), (0,2), (1,3),$ and $(1,4)$.

### 12. Shannon Entropy (in bits)
* **Marginal $H(X)$:**
    $H(X) = -(0.7 \log_2 0.7 + 0.3 \log_2 0.3) \approx 0.881$ bits.
* **Marginal $H(Y)$:**
    $H(Y) = -(0.5 \log_2 0.5 + 0.2 \log_2 0.2 + 0.1 \log_2 0.1 + 0.2 \log_2 0.2) \approx 1.761$ bits.
* **Joint $H(X, Y)$:**
    Since $X$ is determined by $Y$, $H(X, Y) = H(Y) \approx 1.761$ bits.
* **Conditional $H(X|Y)$:**
    $H(X|Y) = 0$ bits (Given $Y$, $X$ is known).
* **Conditional $H(Y|X)$:**
    $H(Y|X) = H(X, Y) - H(X) = 1.761 - 0.881 = 0.880$ bits.

# Q10) Information Available in Continuous Random Varaibles

A chemical plant monitors each production run using two different measurement systems.

For a given run, let

* $T$ be the deviation of the reactor temperature from target, measured in standardized units,
* $P$ be the deviation of the reactor pressure from target, also measured in standardized units.

Assume that $(T,P)$ is jointly continuous, with independent normal components:

$$
T \sim N(0,1), \qquad P \sim N(0,4),
$$

and $T$ and $P$ are independent.

Let the underlying probability space be

$$
\Omega=\mathbb R^2, \qquad \mathfrak F=\mathfrak B(\mathbb R^2),
$$

and let $P$ be the probability measure induced by the joint density

$$
f_{T,P}(t,p)=\frac{1}{4\pi}\exp\left(-\frac{t^2}{2}-\frac{p^2}{8}\right), \qquad (t,p)\in\mathbb R^2.
$$

Define two random variables on the same probability space:

* $X:\Omega\to\mathbb R$ by
  $$
  X(t,p)=t,
  $$
  so that $X$ records only the reactor temperature deviation;

* $Y:\Omega\to\mathbb R^2$ by
  $$
  Y(t,p)=(t,p),
  $$
  so that $Y$ records the full sensor state: both temperature and pressure deviations.

Answer the following.

1. Write down the probability space $(\Omega,\mathfrak F,P)$ explicitly.

2. Compute the $\sigma$-algebra generated by $X$, namely $\sigma(X)$.

3. Compute the $\sigma$-algebra generated by $Y$, namely $\sigma(Y)$.

4. Show that
   $$
   \sigma(X)\subseteq \sigma(Y),
   $$
   and explain why $Y$ contains more information than $X$.

5. Find the marginal density of $X$.

6. Find the marginal density of $Y$.

7. Compute the following probabilities:
   $$
   P(X\leq 0), \qquad P(X>1), \qquad P(Y\in (-\infty,0]\times\mathbb R), \qquad P(Y\in [ -1,1]\times[-2,2]).
   $$

8. Compute the conditional distribution of $X$ given $Y=(t,p)$.

9. Compute the conditional distribution of $Y$ given $X=t$.

10. Explain, in words, what is known when one conditions on $X$ and what additional information becomes available when one conditions on $Y$.

11. Show that
    $$
    X=\pi_1\circ Y,
    $$
    where $\pi_1(t,p)=t$, and explain why this implies that $X$ is a measurable function of $Y$.

12. Give a Python-based visualization of the model, for example by plotting:

* the joint density of $Y=(T,P)$ as a contour plot or heat map;
* the marginal density of $X$;
* and a diagram illustrating that $\sigma(X)$ is coarser than $\sigma(Y)$.

13. Compute the differential entropy of $X$:

$$
h(X)=-\int_{\mathbb R} f_X(x)\ln f_X(x)\,dx.
$$

14. Compute the differential entropy of $Y$:

$$
h(Y)=-\int_{\mathbb R^2} f_Y(t,p)\ln f_Y(t,p)\,dt\,dp.
$$

15. Compute the conditional differential entropy

$$
h(Y|X),
$$

and interpret it physically.

16. Discuss what happens to

$$
h(X|Y)
$$

in this model, and explain why the continuous case differs from the discrete case when one random variable is completely determined by the other.


### 1. Probability Space $(\Omega, \mathfrak{F}, P)$
The probability space is defined as:
* **Sample Space**: $\Omega = \mathbb{R}^2$.
* **$\sigma$-algebra**: $\mathfrak{F} = \mathfrak{B}(\mathbb{R}^2)$, the Borel $\sigma$-algebra on $\mathbb{R}^2$.
* **Probability Measure**: For any $A \in \mathfrak{F}$, $P(A) = \iint_A \frac{1}{4\pi} \exp\left(-\frac{t^2}{2} - \frac{p^2}{8}\right) dt \, dp$.

---

### 2. $\sigma$-algebra generated by $X$
$\sigma(X)$ is the collection of all inverse images of Borel sets in $\mathbb{R}$ under the map $X(t, p) = t$:
$$\sigma(X) = \{ B \times \mathbb{R} : B \in \mathfrak{B}(\mathbb{R}) \}$$
Visually, these are "vertical strips" in the plane $\mathbb{R}^2$.

---

### 3. $\sigma$-algebra generated by $Y$
Since $Y(t, p) = (t, p)$ is the identity map on $\Omega = \mathbb{R}^2$:
$$\sigma(Y) = \mathfrak{B}(\mathbb{R}^2)$$
This contains all Borel subsets of the plane (rectangles, circles, etc.).

---

### 4. Inclusion $\sigma(X) \subseteq \sigma(Y)$
Every set in $\sigma(X)$ is of the form $B \times \mathbb{R}$, which is a Borel set in $\mathbb{R}^2$. Thus, $\sigma(X) \subseteq \sigma(Y)$.
**Information Interpretation**: $Y$ contains "more information" because knowing the outcome of $Y$ tells you both $T$ and $P$, whereas knowing $X$ only tells you $T$. In terms of $\sigma$-algebras, $\sigma(Y)$ is finer; it can distinguish between points $(t, p_1)$ and $(t, p_2)$, while $\sigma(X)$ cannot.

---

### 5. Marginal Density of $X$
$X$ is the temperature deviation $T$. Given $T \sim N(0,1)$:
$$f_X(x) = \frac{1}{\sqrt{2\pi}} e^{-x^2/2}$$

---

### 6. Marginal Density of $Y$
$Y$ is the joint vector $(T, P)$. Its density is the joint density provided:
$$f_Y(t, p) = \frac{1}{4\pi} \exp\left(-\frac{t^2}{2} - \frac{p^2}{8}\right)$$

---

### 7. Probabilities
* **$P(X \leq 0)$**: Since $X \sim N(0,1)$, by symmetry $P(X \leq 0) = 0.5$.
* **$P(X > 1)$**: $1 - \Phi(1) \approx 1 - 0.8413 = 0.1587$.
* **$P(Y \in (-\infty, 0] \times \mathbb{R})$**: This is $P(T \leq 0, P \in \mathbb{R}) = P(T \leq 0) = 0.5$.
* **$P(Y \in [-1, 1] \times [-2, 2])$**: Since $T, P$ are independent:
    $$P(-1 \leq T \leq 1) \cdot P(-2 \leq P \leq 2) = (\Phi(1) - \Phi(-1)) \cdot (\Phi(2/2) - \Phi(-2/2))$$
    $$= (0.6827) \cdot (0.6827) \approx 0.4661$$

---

### 8. Conditional Distribution of $X$ given $Y=(t, p)$
Since $X$ is a component of $Y$ ($X=t$ when $Y=(t,p)$), the distribution is a **Dirac delta** (degenerate):
$$X | Y=(t,p) \sim \delta_t$$
There is no uncertainty in $X$ once $Y$ is known.

---

### 9. Conditional Distribution of $Y$ given $X=t$
$Y = (T, P)$. Given $X=t$, we know $T=t$ with certainty. Because $T$ and $P$ are independent, the distribution of $P$ remains $N(0, 4)$.
$$Y | X=t \sim \delta_t \times N(0, 4)$$

---

### 10. Information Interpretation
* **Conditioning on $X$**: We know the exact temperature deviation. The pressure deviation remains completely unknown (described by its original marginal distribution).
* **Conditioning on $Y$**: We know both temperature and pressure. No uncertainty remains regarding the sensor state.

---

### 11. $X$ as a Measurable Function of $Y$
$X(t, p) = t$ and $Y(t, p) = (t, p)$, so $X = \pi_1(Y)$ where $\pi_1$ is the projection map. Since $\pi_1$ is continuous, it is Borel measurable. By the Doob-Dynkin Lemma, if $X = f(Y)$ for a measurable $f$, then $X$ is $\sigma(Y)$-measurable, implying $\sigma(X) \subseteq \sigma(Y)$.

---

### 13. Differential Entropy $h(X)$
For a normal distribution $N(\mu, \sigma^2)$, $h = \frac{1}{2} \ln(2\pi e \sigma^2)$.
For $X \sim N(0, 1)$:
$$h(X) = \frac{1}{2} \ln(2\pi e) \approx 1.4189 \text{ nats}$$

---

### 14. Differential Entropy $h(Y)$
Since $T$ and $P$ are independent, $h(Y) = h(T) + h(P)$.
* $h(T) = \frac{1}{2} \ln(2\pi e (1))$
* $h(P) = \frac{1}{2} \ln(2\pi e (4))$
$$h(Y) = \ln(2\pi e \cdot 2) = \ln(4\pi e) \approx 3.528 \text{ nats}$$

---

### 15. Conditional Differential Entropy $h(Y|X)$
Because $T$ and $P$ are independent, knowing $X$ (which is $T$) only reduces the uncertainty of the $T$ component to zero.
$$h(Y|X) = h(T, P | T) = h(T|T) + h(P|T) = 0 + h(P)$$
$$h(Y|X) = \frac{1}{2} \ln(2\pi e \cdot 4) = \frac{1}{2} \ln(8\pi e) \approx 2.11 \text{ nats}$$
**Physical Interpretation**: This represents the remaining uncertainty in the system (the pressure) after the temperature has been measured.

---

### 16. The $h(X|Y)$ Paradox
In this model, $X$ is a deterministic function of $Y$.
* In **discrete** entropy, $H(X|Y) = 0$ if $X = f(Y)$.
* In **continuous** differential entropy, $h(X|Y)$ technically approaches $-\infty$.

Differential entropy does not behave like discrete entropy regarding functional dependence; while $X$ has no "uncertainty" given $Y$, the differential entropy reflects the "volume" of the support. A point mass has zero volume, leading to a logarithm of zero, i.e., $-\infty$.

\Omega=\{\omega_1,\omega_2,\omega_3,\omega_4\},

* $\omega_4$: high-risk customer, fraudulent transaction.

$$

P(\{\omega_2\})=0.10,\qquad

$$

Let $X:\Omega\to\mathbb R$ denote the fraud indicator:

$$
X(\omega_1)=0,\qquad
X(\omega_2)=1,\qquad
X(\omega_3)=0,\qquad

Now suppose the bank does not observe the full outcome $\omega$, but only the customer’s risk class. Define $Y:\Omega\to\mathbb R$ by

$$
Y(\omega_1)=0,\qquad
Y(\omega_2)=0,\qquad
Y(\omega_3)=1,\qquad

* $Y=0$ means “low-risk customer,”
* $Y=1$ means “high-risk customer.”

3. Explain why $E[X\mid Y]$ must be $\sigma(Y)$-measurable, and therefore must be constant on the sets ${ \omega_1,\omega_2}$ and ${\omega_3,\omega_4}$.

4. Compute $E[X\mid Y]$ explicitly.

\int_A E[X\mid Y]\,dP=\int_A X\,dP.

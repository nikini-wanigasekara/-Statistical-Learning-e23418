$$
x_1,x_2,\dots,x_n\in\mathbb R^d.
$$
We wish to cluster these observations into $K$ groups. Instead of assigning each point deterministically to a cluster at the beginning, we introduce a latent random variable
$$
C_i\in{1,\dots,K},

$$
P(C_i=k)=\phi_k,

$$
\phi_k\ge 0,

$$

$$

\qquad
\Sigma_k\in\mathbb R^{d\times d}

# Answer : Gaussian Mixture Clustering as Conditional Updating

### 1. Deriving the Marginal Density

By the law of total probability, partitioning the sample space by the mutually exclusive, exhaustive events $\{C_i=k\}_{k=1}^K$:

$$
p(x_i)=\sum_{k=1}^K P(X_i=x_i, C_i=k)=\sum_{k=1}^K P(X_i=x_i\mid C_i=k)P(C_i=k)
$$

Substituting the Gaussian conditional density and the prior $\phi_k$:

$$
p(x_i)=\sum_{k=1}^K \phi_k\,\mathscr N(x_i\mid \mu_k,\Sigma_k)
$$

This is called a **Gaussian mixture density** because it is a convex combination (weights $\phi_k\ge0$, $\sum_k\phi_k=1$) of $K$ Gaussian component densities — the overall density is a weighted "mixture" of several Gaussian populations rather than a single Gaussian.

---

### 2. Deriving the Posterior Cluster Probability

By Bayes' rule, for events $C_i=k$ and $X_i=x_i$:

$$
P(C_i=k\mid X_i=x_i)=\frac{P(X_i=x_i\mid C_i=k)\,P(C_i=k)}{P(X_i=x_i)}=\frac{P(X_i=x_i\mid C_i=k)P(C_i=k)}{\sum_{j=1}^K P(X_i=x_i\mid C_i=j)P(C_i=j)}
$$

since the denominator is $p(x_i)$ expanded via the law of total probability (Part 1). Substituting $P(X_i=x_i\mid C_i=k)=\mathscr N(x_i\mid\mu_k,\Sigma_k)$ and $P(C_i=k)=\phi_k$:

$$
\gamma_{ik}:=P(C_i=k\mid X_i=x_i)=\frac{\phi_k\,\mathscr N(x_i\mid \mu_k,\Sigma_k)}{\sum_{j=1}^K \phi_j\,\mathscr N(x_i\mid \mu_j,\Sigma_j)}
$$

$\gamma_{ik}$ is a **posterior probability of cluster membership** because it updates the *prior* belief $\phi_k$ (before seeing $x_i$) into a *posterior* belief conditional on the evidence $X_i=x_i$, using exactly the Bayesian updating mechanism: prior × likelihood, normalized.

---

### 3. One-Hot Encoding of the Latent Cluster Variable

$Z_{ik}$ is a Bernoulli-type indicator: $Z_{ik}=\mathbb 1\{C_i=k\}$. For any indicator random variable, $\mathbb E[\mathbb 1\{A\}\mid \mathcal F]=P(A\mid \mathcal F)$. Hence:

$$
\mathbb E[Z_{ik}\mid X_i=x_i]=1\cdot P(C_i=k\mid X_i=x_i)+0\cdot P(C_i\ne k\mid X_i=x_i)=P(C_i=k\mid X_i=x_i)=\gamma_{ik}
$$

Stacking over $k=1,\dots,K$ (expectation acts componentwise on a random vector):

$$
\mathbb E[Z_i\mid X_i=x_i]=\begin{bmatrix}\mathbb E[Z_{i1}\mid X_i=x_i]\\ \vdots\\ \mathbb E[Z_{iK}\mid X_i=x_i]\end{bmatrix}=\begin{bmatrix}\gamma_{i1}\\ \vdots\\ \gamma_{iK}\end{bmatrix}
$$

Since $Z_i$ is a one-hot encoding of cluster membership, the vector of soft assignment weights used in GMM clustering is **exactly** the conditional expectation $\mathbb E[Z_i\mid X_i=x_i]$ — not an ad hoc heuristic but a direct probabilistic quantity.

---

### 4. From Soft Assignment to Hard Clustering

$\mathbb E[Z_i\mid X_i=x_i]=(\gamma_{i1},\dots,\gamma_{iK})$ is a **soft assignment**: it distributes a total probability mass of 1 fractionally across all $K$ clusters, reflecting genuine uncertainty about which cluster generated $x_i$ (e.g. a point near a cluster boundary might get $\gamma_{i1}=0.55,\gamma_{i2}=0.45$).

**Hard clustering** collapses this distribution to a single deterministic label by taking the mode:

$$
\widehat C_i=\arg\max_{1\le k\le K}\gamma_{ik}
$$

This discards the uncertainty information — two points with responsibilities $(0.99,0.01)$ and $(0.51,0.49)$ get the same hard label even though the second is far more ambiguous.

---

### 5. Conditional Expectation of the Observation Given the Cluster

Since $X_i\mid C_i=k\sim \mathscr N(\mu_k,\Sigma_k)$, and the mean of a multivariate Gaussian $\mathscr N(\mu_k,\Sigma_k)$ is $\mu_k$ by definition of the distribution's first moment:

$$
\mathbb E[X_i\mid C_i=k]=\mu_k
$$

$\mu_k$ is the **center of cluster $k$** because it is literally the expected location of any observation known to originate from component $k$ — the "average point" generated by that cluster's Gaussian.

**Comparison:**
- $\mathbb E[Z_i\mid X_i=x_i]$ is a conditional expectation of the *latent label* given an *observed point* — it answers "given this data point, which cluster does it likely belong to?" (soft cluster membership of an observation).
- $\mathbb E[X_i\mid C_i=k]$ is a conditional expectation of the *observation* given a *fixed latent cluster* — it answers "if I know a point came from cluster $k$, where do I expect it to be?" (the mean location/center of a cluster).

These are conditional expectations taken in opposite directions along the same generative model $C_i \to X_i$.

---

### 6. The Complete-Data Likelihood

Given labels $z_i$, since $z_{ik}\in\{0,1\}$ and exactly one $z_{ik}=1$ per $i$:

$$
p(x_1,\dots,x_n,z_1,\dots,z_n)=\prod_{i=1}^n\prod_{k=1}^K\big[\phi_k\,\mathscr N(x_i\mid\mu_k,\Sigma_k)\big]^{z_{ik}}
$$

Taking logs and using $\log(\prod)=\sum\log$, and $\log(a^{z})=z\log a$:

$$
\ell_c=\log p(\cdot)=\sum_{i=1}^n\sum_{k=1}^K z_{ik}\Big[\log\phi_k+\log \mathscr N(x_i\mid\mu_k,\Sigma_k)\Big]
$$

If the $z_{ik}$ were known, this expression would be **easy to maximize** because the indicator $z_{ik}$ turns the sum over $k$ into a *selection* — for each $i$, only the true-cluster term survives. The optimization decouples completely by cluster: maximizing over $(\phi_k,\mu_k,\Sigma_k)$ reduces to fitting a single Gaussian's parameters to the subset of points labeled $k$, and $\phi_k$ to simple class-proportion counting — exactly like the complete-data MLE for a labeled Gaussian classification problem, no coupling between clusters remains.

---

### 7. The EM Interpretation

Since $z_{ik}$ is unobserved, EM replaces it by its conditional expectation given the data and current parameters (Part 3): $z_{ik}\rightsquigarrow \mathbb E[Z_{ik}\mid X_i=x_i]=\gamma_{ik}$. Substituting into $\ell_c$ gives the **expected complete-data log-likelihood**:

$$
Q=\sum_{i=1}^n\sum_{k=1}^K \gamma_{ik}\Big[\log\phi_k+\log\mathscr N(x_i\mid\mu_k,\Sigma_k)\Big]
$$

This is precisely $\mathbb E_{Z\mid X,\theta^{\text{old}}}[\ell_c]$, i.e. taking expectation of the complete-data log-likelihood over the posterior distribution of the latent labels given current parameter estimates. The **E-step is a conditional update of cluster membership probabilities** because it recomputes $\gamma_{ik}=P(C_i=k\mid X_i=x_i,\theta^{\text{old}})$ under the current parameter values — it is literally Bayesian posterior updating of the latent cluster variable, refreshed every iteration as the parameters change.

---

### 8. Parameter Updates

Maximizing $Q$ subject to $\sum_k \phi_k=1$ (via a Lagrange multiplier) and setting $\partial Q/\partial \mu_k=0$, $\partial Q/\partial \Sigma_k=0$ yields the standard closed-form M-step (each term in $Q$ separates by $k$, and only the Gaussian log-density depends on $\mu_k,\Sigma_k$, giving weighted Gaussian MLE equations):

$$
N_k=\sum_{i=1}^n \gamma_{ik},\qquad
\phi_k^{\text{new}}=\frac{N_k}{n},\qquad
\mu_k^{\text{new}}=\frac{1}{N_k}\sum_{i=1}^n \gamma_{ik}x_i,\qquad
\Sigma_k^{\text{new}}=\frac{1}{N_k}\sum_{i=1}^n \gamma_{ik}(x_i-\mu_k^{\text{new}})(x_i-\mu_k^{\text{new}})^T
$$

$\gamma_{ik}$ acts as a **fractional membership weight**: instead of each point contributing a full unit of "count" to exactly one cluster (as in hard k-means), point $i$ contributes $\gamma_{ik}$ units of weight to cluster $k$'s mean, covariance, and effective sample size $N_k$ — a point split 70/30 between two clusters contributes 0.7 of itself to one cluster's statistics and 0.3 to the other's, so no observation is "wasted" or arbitrarily discretized.

---

### 9. Interpretation

Gaussian mixture clustering can be viewed as a repeated process of conditional updating because every quantity in the model is defined through conditioning. The mixture weight $\phi_k$ encodes the prior probability that any given observation belongs to cluster $k$, before any data is examined. Once an observation $x_i$ is observed, the Gaussian density $\mathscr N(x_i\mid\mu_k,\Sigma_k)$ measures how compatible that point is with each cluster's location and spread — it plays the role of the likelihood. Combining the prior $\phi_k$ with this likelihood via Bayes' rule produces the responsibility $\gamma_{ik}$, the posterior probability that cluster $k$ generated $x_i$ after the evidence has been taken into account. Collecting these responsibilities across all $k$ gives the soft assignment vector $\mathbb E[Z_i\mid X_i=x_i]$, the conditional expectation of the one-hot cluster indicator given the data. The M-step then updates $\phi_k,\mu_k,\Sigma_k$ using these posterior probabilities as weights, effectively re-estimating the model's beliefs about cluster shape and prevalence from the newly updated membership probabilities. Iterating E- and M-steps therefore is nothing but an alternating sequence of Bayesian conditional updates: given current parameters, update the latent membership probabilities; given the updated memberships, update the parameters. Gaussian mixture clustering is thus probabilistic clustering based on conditional expectations of latent cluster membership variables, not a heuristic partitioning rule.

```python
!pip install kagglehub -q

import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.mixture import GaussianMixture
import plotly.graph_objects as go


class GMMFinancialSegmenter:
    """
    Two-dimensional Gaussian Mixture Model segmenter with Plotly diagnostics,
    illustrating soft cluster assignment as the conditional expectation
    E[Z_i | X_i = x] derived analytically in Part 3.
    """

    def __init__(self, n_components=3, random_state=42):
        self.n_components = n_components
        self.random_state = random_state
        self.scaler = StandardScaler()
        self.gmm = GaussianMixture(
            n_components=self.n_components,
            covariance_type='full',
            random_state=self.random_state
        )

    def load_and_prepare(self, df, feature_cols=('PURCHASES', 'CREDIT_LIMIT'), test_size=0.2):
        data = df[list(feature_cols)].dropna().copy()
        self.feature_cols = list(feature_cols)
        X = data.values
        X_train, X_test = train_test_split(
            X, test_size=test_size, random_state=self.random_state
        )
        self.X_train_raw, self.X_test_raw = X_train, X_test
        self.X_train = self.scaler.fit_transform(X_train)
        self.X_test = self.scaler.transform(X_test)
        return self.X_train, self.X_test

    def fit(self):
        self.gmm.fit(self.X_train)
        print(f"Converged: {self.gmm.converged_}")
        print(f"Iterations to converge: {self.gmm.n_iter_}")
        return self.gmm

    def evaluate_out_of_sample(self):
        avg_ll = self.gmm.score(self.X_test)
        print(f"Average out-of-sample log-likelihood: {avg_ll:.4f}")
        return avg_ll

    def plot_density_heatmap(self):
        fig = go.Figure()
        fig.add_trace(go.Histogram2d(
            x=self.X_train[:, 0], y=self.X_train[:, 1],
            colorscale='Viridis', nbinsx=40, nbinsy=40
        ))
        fig.update_layout(
            title="Empirical 2D Density Heatmap of Training Data",
            xaxis_title=f"{self.feature_cols[0]} (scaled)",
            yaxis_title=f"{self.feature_cols[1]} (scaled)"
        )
        fig.show()
        return fig

    def _responsibility_grid(self, X, grid_res=150):
        x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
        y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
        xx, yy = np.meshgrid(
            np.linspace(x_min, x_max, grid_res),
            np.linspace(y_min, y_max, grid_res)
        )
        grid_points = np.column_stack([xx.ravel(), yy.ravel()])
        resp = self.gmm.predict_proba(grid_points)          # E[Z_i | X_i = x_grid]
        max_resp = resp.max(axis=1).reshape(xx.shape)
        return xx, yy, max_resp

    def _assignment_plot(self, X, title):
        xx, yy, max_resp = self._responsibility_grid(X)
        labels = self.gmm.predict(X)
        fig = go.Figure()
        fig.add_trace(go.Contour(
            x=xx[0], y=yy[:, 0], z=max_resp,
            colorscale='Viridis', opacity=0.55,
            contours=dict(showlines=False),
            colorbar=dict(title="max γ_ik")
        ))
        fig.add_trace(go.Scatter(
            x=X[:, 0], y=X[:, 1], mode='markers',
            marker=dict(color=labels, colorscale='Turbo', size=6,
                        line=dict(width=0.5, color='black')),
            name='data'
        ))
        fig.update_layout(
            title=title,
            xaxis_title=f"{self.feature_cols[0]} (scaled)",
            yaxis_title=f"{self.feature_cols[1]} (scaled)"
        )
        fig.show()
        return fig

    def plot_training_assignments(self):
        return self._assignment_plot(self.X_train, "Training Assignment Plot (max responsibility contours)")

    def plot_test_assignments(self):
        return self._assignment_plot(self.X_test, "Test Assignment Plot (out-of-sample)")


# ---------------------------------------------------------------------------
# Load the CC General dataset (Kaggle: arjunbhasin2013/ccdata)
# ---------------------------------------------------------------------------
import kagglehub

path = kagglehub.dataset_download("arjunbhasin2013/ccdata")
df = pd.read_csv(f"{path}/CC GENERAL.csv")

# ---------------------------------------------------------------------------
# Run the full pipeline
# ---------------------------------------------------------------------------
segmenter = GMMFinancialSegmenter(n_components=3, random_state=42)
segmenter.load_and_prepare(df, feature_cols=('PURCHASES', 'CREDIT_LIMIT'))
segmenter.fit()
segmenter.evaluate_out_of_sample()

segmenter.plot_density_heatmap()
segmenter.plot_training_assignments()
segmenter.plot_test_assignments()
```

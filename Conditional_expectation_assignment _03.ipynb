#Q1) Prediction of fraudulent transactions

A bank monitors credit-card transactions for possible fraud.

For each transaction, let the sample space be

$$
\Omega=\{\omega_1,\omega_2,\omega_3,\omega_4\},
$$

where:

* $\omega_1$: low-risk customer, legitimate transaction,
* $\omega_2$: low-risk customer, fraudulent transaction,
* $\omega_3$: high-risk customer, legitimate transaction,
* $\omega_4$: high-risk customer, fraudulent transaction.

Assume the probabilities are

$$
P(\{\omega_1\})=0.50,\qquad
P(\{\omega_2\})=0.10,\qquad
P(\{\omega_3\})=0.20,\qquad
P(\{\omega_4\})=0.20.
$$

Let $X:\Omega\to\mathbb R$ denote the fraud indicator:

$$
X(\omega_1)=0,\qquad
X(\omega_2)=1,\qquad
X(\omega_3)=0,\qquad
X(\omega_4)=1.
$$

Thus $X=1$ means the transaction is fraudulent, and $X=0$ means it is legitimate.

Now suppose the bank does not observe the full outcome $\omega$, but only the customer's risk class. Define $Y:\Omega\to\mathbb R$ by

$$
Y(\omega_1)=0,\qquad
Y(\omega_2)=0,\qquad
Y(\omega_3)=1,\qquad
Y(\omega_4)=1.
$$

Thus:

* $Y=0$ means “low-risk customer,”
* $Y=1$ means “high-risk customer.”

Answer the following.

1. Write down the probability space $(\Omega,\mathfrak F,P)$, where $\mathfrak F=\mathfrak P(\Omega)$.

2. Compute the $\sigma$-algebra generated by $Y$, namely $\sigma(Y)$.

3. Explain why $E[X\mid Y]$ must be $\sigma(Y)$-measurable, and therefore must be constant on the sets $\{ \omega_1,\omega_2\}$ and $\{\omega_3,\omega_4\}$.

4. Compute $E[X\mid Y]$ explicitly and show that

$$
E[X\mid Y]= \frac{1}{6} \mathbb{1}_{\{Y=0\}} + \frac{1}{2} \mathbb{1}_{\{Y=1\}}
$$

5. Verify the defining property of conditional expectation by checking that for every set $A\in\sigma(Y)$,

$$
\int_A E[X\mid Y]\,dP=\int_A X\,dP.
$$

6. Find the Conditional PMF $p_{X|Y}(x \mid y)$.

7. Verify the identity:
$$\mathbb{E}[X\mid Y=y] = \sum_{x \in \{0,1\}} x \cdot p_{X|Y}(x \mid y)$$.

8. Calculate the **Mean Square Error (MSE)**

9. How would the **Mean Square Error (MSE)** change if the bank decided to ignore the customer risk class $Y$ and instead used the global average fraud rate ($E[X] = 0.30$) as a constant estimate for every transaction? Calculate this error and explain its relationship to the **Law of Total Variance**.

10. Using the joint distribution of $(X, Y)$ derived from the sample space:

    a) Compute the entropy $H(X)$, representing the bank's initial uncertainty regarding fraud.

    b) Compute the conditional entropy $H(X \mid Y)$, representing the uncertainty remaining after the customer's risk class is revealed.

    c) Calculate the Mutual Information $I(X; Y)$.

    d) Based on these values, what percentage of the total uncertainty about fraud is removed by knowing the customer's risk class?

## Answers for Q1) :

## 1.

The probability space is a triplet $(\Omega,\mathfrak F,P)$ defined as follows:

* **Sample Space:** $\Omega=\{\omega_1,\omega_2,\omega_3,\omega_4\}$
* **$\sigma$-algebra:** $\mathfrak F$ is the power set of $\Omega$, denoted $\mathfrak P(\Omega)$, which contains all $2^4 = 16$ possible subsets of $\Omega$:

$$\mathfrak F = \big\{ \emptyset, \{\omega_1\}, \{\omega_2\}, \{\omega_3\}, \{\omega_4\}, \{\omega_1,\omega_2\}, \{\omega_1,\omega_3\}, \{\omega_1,\omega_4\}, \{\omega_2,\omega_3\}, \{\omega_2,\omega_4\}, \{\omega_3,\omega_4\}, \{\omega_1,\omega_2,\omega_3\}, \{\omega_1,\omega_2,\omega_4\}, \{\omega_1,\omega_3,\omega_4\}, \{\omega_2,\omega_3,\omega_4\}, \Omega \big\}$$

* **Probability Measure:** $P:\mathfrak F \to [0,1]$ is uniquely determined by the probabilities assigned to the singletons, extended by additivity to all sets in $\mathfrak F$:

$$P(\{\omega_1\})=0.50,\qquad
P(\{\omega_2\})=0.10,\qquad
P(\{\omega_3\})=0.20,\qquad
P(\{\omega_4\})=0.20.$$

---

## 2.

The random variable $Y$ induces a partition on $\Omega$ based on its pre-images:

* For $Y=0$: $Y^{-1}(\{0\}) = \{\omega_1, \omega_2\}$
* For $Y=1$: $Y^{-1}(\{1\}) = \{\omega_3, \omega_4\}$

The $\sigma$-algebra $\sigma(Y)$ is the collection of all sets generated by these inverse images, which includes the empty set, the partition sets, and their unions:

$$\sigma(Y) = \big\{ \emptyset, \{\omega_1, \omega_2\}, \{\omega_3, \omega_4\}, \Omega \big\}$$

---

## 3.

By definition, the conditional expectation $E[X \mid Y]$ is a random variable that is a function of $Y$ alone. For it to be a valid random variable with respect to the information revealed by $Y$, it must be **$\sigma(Y)$-measurable**.

The atoms (the smallest non-empty building blocks) of the $\sigma$-algebra $\sigma(Y)$ are exactly the sets $A_0 = \{\omega_1, \omega_2\}$ and $A_1 = \{\omega_3, \omega_4\}$. For any function to be measurable with respect to a $\sigma$-algebra, it must be constant over the atoms of that $\sigma$-algebra. If it took different values for elements within the same atom, resolving the function would require more granular information than what $\sigma(Y)$ provides, violating measurability. Therefore, $E[X\mid Y]$ must be constant on $\{\omega_1,\omega_2\}$ and constant on $\{\omega_3,\omega_4\}$.

---

## 4.

Since $E[X \mid Y]$ is constant on the sets $\{Y=0\}$ and $\{Y=1\}$, we compute its value on each partition set using the conditional expectation formula:

* **On the set $\{Y=0\} = \{\omega_1, \omega_2\}$:**

$$E[X \mid Y=0] = \frac{E[X \cdot \mathbb{1}_{\{\omega_1,\omega_2\}}]}{P(\{\omega_1,\omega_2\})} = \frac{X(\omega_1)P(\{\omega_1\}) + X(\omega_2)P(\{\omega_2\})}{P(\{\omega_1\}) + P(\{\omega_2\})}$$

$$E[X \mid Y=0] = \frac{0(0.50) + 1(0.10)}{0.50 + 0.10} = \frac{0.10}{0.60} = \frac{1}{6}$$

* **On the set $\{Y=1\} = \{\omega_3, \omega_4\}$:**

$$E[X \mid Y=1] = \frac{E[X \cdot \mathbb{1}_{\{\omega_3,\omega_4\}}]}{P(\{\omega_3,\omega_4\})} = \frac{X(\omega_3)P(\{\omega_3\}) + X(\omega_4)P(\{\omega_4\})}{P(\{\omega_3\}) + P(\{\omega_4\})}$$

$$E[X \mid Y=1] = \frac{0(0.20) + 1(0.20)}{0.20 + 0.20} = \frac{0.20}{0.40} = \frac{1}{2}$$

Combining these using indicator functions yields:

$$E[X\mid Y]= \frac{1}{6} \mathbb{1}_{\{Y=0\}} + \frac{1}{2} \mathbb{1}_{\{Y=1\}}$$

---

## 5.

We test this identity across all four sets in $\sigma(Y) = \big\{ \emptyset, \{\omega_1, \omega_2\}, \{\omega_3, \omega_4\}, \Omega \big\}$:

* **Case 1: $A = \emptyset$**

$$\int_{\emptyset} E[X\mid Y]\,dP = 0 = \int_{\emptyset} X\,dP \quad \checkmark$$

* **Case 2: $A = \{\omega_1, \omega_2\}$**

$$\int_{\%\{\omega_1,\omega_2\}} E[X\mid Y]\,dP = \frac{1}{6} \cdot P(\{\omega_1,\omega_2\}) = \frac{1}{6} \cdot 0.60 = 0.10$$

$$\int_{\{\omega_1,\omega_2\}} X\,dP = X(\omega_1)P(\{\omega_1\}) + X(\omega_2)P(\{\omega_2\}) = 0(0.50) + 1(0.10) = 0.10 \quad \checkmark$$

* **Case 3: $A = \{\omega_3, \omega_4\}$**

$$\int_{\{\omega_3,\omega_4\}} E[X\mid Y]\,dP = \frac{1}{2} \cdot P(\{\omega_3,\omega_4\}) = \frac{1}{2} \cdot 0.40 = 0.20$$

$$\int_{\{\omega_3,\omega_4\}} X\,dP = X(\omega_3)P(\{\omega_3\}) + X(\omega_4)P(\{\omega_4\}) = 0(0.20) + 1(0.20) = 0.20 \quad \checkmark$$

* **Case 4: $A = \Omega$**

$$\int_{\Omega} E[X\mid Y]\,dP = \frac{1}{6}(0.60) + \frac{1}{2}(0.40) = 0.10 + 0.20 = 0.30$$

$$\int_{\Omega} X\,dP = \sum_{i=1}^4 X(\omega_i)P(\{\omega_i\}) = 0(0.50) + 1(0.10) + 0(0.20) + 1(0.20) = 0.30 \quad \checkmark$$

The identity holds for all $A \in \sigma(Y)$.

---

## 6.

The conditional probability mass function is given by $p_{X|Y}(x \mid y) = \frac{P(X=x, Y=y)}{P(Y=y)}$.

* **For $y = 0$ ($P(Y=0) = 0.60$):**
* $p_{X|Y}(0 \mid 0) = \frac{P(\{\omega_1\})}{P(Y=0)} = \frac{0.50}{0.60} = \frac{5}{6}$
* $p_{X|Y}(1 \mid 0) = \frac{P(\{\omega_2\})}{P(Y=0)} = \frac{0.10}{0.60} = \frac{1}{6}$


* **For $y = 1$ ($P(Y=1) = 0.40$):**
* $p_{X|Y}(0 \mid 1) = \frac{P(\{\omega_3\})}{P(Y=1)} = \frac{0.20}{0.40} = \frac{1}{2}$
* $p_{X|Y}(1 \mid 1) = \frac{P(\{\omega_4\})}{P(Y=1)} = \frac{0.20}{0.40} = \frac{1}{2}$



### Summary Table: $p_{X|Y}(x \mid y)$

| $x \setminus y$ | $y = 0$ | $y = 1$ |
| --- | --- | --- |
| **$x = 0$** | $5/6$ | $1/2$ |
| **$x = 1$** | $1/6$ | $1/2$ |

---

## 7.

We evaluate the right-hand side expression for both values of $y$:

* **For $y = 0$:**

$$\sum_{x \in \{0,1\}} x \cdot p_{X|Y}(x \mid 0) = 0 \cdot p_{X|Y}(0 \mid 0) + 1 \cdot p_{X|Y}(1 \mid 0) = 0\left(\frac{5}{6}\right) + 1\left(\frac{1}{6}\right) = \frac{1}{6}$$

This matches $E[X \mid Y=0] = \frac{1}{6}$ calculated in Question 4.

* **For $y = 1$:**

$$\sum_{x \in \{0,1\}} x \cdot p_{X|Y}(x \mid 1) = 0 \cdot p_{X|Y}(0 \mid 1) + 1 \cdot p_{X|Y}(1 \mid 1) = 0\left(\frac{1}{2}\right) + 1\left(\frac{1}{2}\right) = \frac{1}{2}$$

This matches $E[X \mid Y=1] = \frac{1}{2}$ calculated in Question 4.

The identity is verified.

---

## 8.

The Mean Square Error when predicting $X$ using $E[X \mid Y]$ is given by:

$$\text{MSE} = E\left[ (X - E[X\mid Y])^2 \right] = \sum_{i=1}^4 \left( X(\omega_i) - E[X\mid Y](\omega_i) \right)^2 P(\{\omega_i\})$$

Mapping values for each outcome $\omega_i$:

* **$\omega_1$**: $\left(0 - \frac{1}{6}\right)^2 (0.50) = \frac{1}{36} \cdot \frac{1}{2} = \frac{1}{72}$
* **$\omega_2$**: $\left(1 - \frac{1}{6}\right)^2 (0.10) = \frac{25}{36} \cdot \frac{1}{10} = \frac{25}{360} = \frac{5}{72}$
* **$\omega_3$**: $\left(0 - \frac{1}{2}\right)^2 (0.20) = \frac{1}{4} \cdot \frac{1}{5} = \frac{1}{20}$
* **$\omega_4$**: $\left(1 - \frac{1}{2}\right)^2 (0.20) = \frac{1}{4} \cdot \frac{1}{5} = \frac{1}{20}$

Summing these individual errors:

$$\text{MSE} = \left(\frac{1}{72} + \frac{5}{72}\right) + \left(\frac{1}{20} + \frac{1}{20}\right) = \frac{6}{72} + \frac{2}{20} = \frac{1}{12} + \frac{1}{10} = \frac{5 + 6}{60} = \frac{11}{60} \approx 0.1833$$

---

## 9.

If the bank uses the constant $E[X] = 0.30$, the new error is the total variance of $X$, since $\text{MSE}_{\text{constant}} = E[(X - E[X])^2] = \text{Var}(X)$.
Since $X$ is a Bernoulli random variable with parameter $p = 0.30$:

$$\text{Var}(X) = p(1-p) = 0.30 \times 0.70 = 0.21$$

### Relationship to the Law of Total Variance

The Law of Total Variance states:

$$\text{Var}(X) = E[\text{Var}(X \mid Y)] + \text{Var}(E[X \mid Y])$$

Where:

* $E[\text{Var}(X \mid Y)]$ is the residual variation left unexplained by $Y$, which corresponds exactly to the optimal $\text{MSE}$ calculated in Question 8 ($\frac{11}{60} \approx 0.1833$).
* $\text{Var}(E[X \mid Y])$ is the variance explained by knowing $Y$.

Let's calculate the variance of the predictions $\text{Var}(E[X \mid Y])$:

$$E[E[X\mid Y]^2] = \left(\frac{1}{6}\right)^2(0.60) + \left(\frac{1}{2}\right)^2(0.40) = \frac{1}{36}\left(\frac{3}{5}\right) + \frac{1}{4}\left(\frac{2}{5}\right) = \frac{1}{60} + \frac{6}{60} = \frac{7}{60}$$

$$\text{Var}(E[X \mid Y]) = E[E[X\mid Y]^2] - (E[X])^2 = \frac{7}{60} - (0.30)^2 = \frac{7}{60} - \frac{9}{100} = \frac{35 - 27}{300} = \frac{8}{300} = \frac{2}{75} \approx 0.0267$$

Checking the decomposition:

$$\text{Var}(X) = \frac{11}{60} + \frac{2}{75} = \frac{55 + 8}{300} = \frac{63}{300} = 0.21$$

**Conclusion:** Ignoring $Y$ increases the MSE from $0.1833$ to $0.21$. The error increases exactly by $\text{Var}(E[X \mid Y]) = 0.0267$, which is the predictive information value lost by dropping the risk classification.

---

## 10.

### a)

$X$ has a marginal distribution of $P(X=1)=0.30$ and $P(X=0)=0.70$. Using base-2 logarithms:

$$H(X) = -0.70 \log_2(0.70) - 0.30 \log_2(0.30)$$

$$H(X) \approx -(0.70 \times -0.5146) - (0.30 \times -1.7370) \approx 0.3602 + 0.5211 = 0.8813 \text{ bits}$$

### b)

$$H(X \mid Y) = P(Y=0)H(X \mid Y=0) + P(Y=1)H(X \mid Y=1)$$

* For $Y=0$: $p_{X|Y}(\cdot \mid 0) = \{\frac{5}{6}, \frac{1}{6}\}$

$$H(X \mid Y=0) = -\frac{5}{6}\log_2\left(\frac{5}{6}\right) - \frac{1}{6}\log_2\left(\frac{1}{6}\right) \approx 0.2183 + 0.4308 = 0.6500 \text{ bits}$$

* For $Y=1$: $p_{X|Y}(\cdot \mid 1) = \{\frac{1}{2}, \frac{1}{2}\}$

$$H(X \mid Y=1) = -\frac{1}{2}\log_2\left(\frac{1}{2}\right) - \frac{1}{2}\log_2\left(\frac{1}{2}\right) = 1.0000 \text{ bit}$$

Combining them:

$$H(X \mid Y) = 0.60(0.6500) + 0.40(1.0000) = 0.3900 + 0.4000 = 0.7900 \text{ bits}$$

### c)

Mutual information measures the information shared between the two variables:

$$I(X; Y) = H(X) - H(X \mid Y) = 0.8813 - 0.7900 = 0.0913 \text{ bits}$$

### d)

The percentage of uncertainty removed is the ratio of mutual information to initial entropy:

$$\text{Uncertainty Removed} = \frac{I(X;Y)}{H(X)} \times 100\% = \frac{0.0913}{0.8813} \times 100\% \approx 10.36\%$$

Knowing the customer's risk class resolves approximately **10.36%** of the bank's initial uncertainty regarding whether a transaction is fraudulent.

# Q2) Conditional Density of the Multivariate Gaussian

Let $\mathbf{Z}$ be the concatenated random matrix
$\mathbf{Z} = \begin{bmatrix} \mathbf{X} \\ \mathbf{Y} \end{bmatrix} \in \mathbb{R}^{n+m}$.


The random variable $\mathbf{Z}$ is said to have a **joint multivariate normal distribution** if every linear combination of its components follows a univariate normal distribution.

For $\mathbf{Z}$ to possess a **joint density function** (meaning its distribution is absolutely continuous with respect to the Lebesgue measure on $\mathbb{R}^{n+m}$), its covariance matrix must be non-singular (positive definite).


The joint density function $f_{\mathbf{X}, \mathbf{Y}}(\mathbf{z})$ for $\mathbf{z} \in \mathbb{R}^{n+m}$ is given by:

$$f_{\mathbf{X}, \mathbf{Y}}(\mathbf{z}) = \frac{1}{\sqrt{(2\pi)^{n+m} |\boldsymbol{\Sigma}|}} \exp\left( -\frac{1}{2} (\mathbf{z} - \boldsymbol{\mu})^T \boldsymbol{\Sigma}^{-1} (\mathbf{z} - \boldsymbol{\mu}) \right).$$

Here:

* **Mean ($\boldsymbol{\mu}$):** $\boldsymbol{\mu} = E[\mathbf{Z}] = \begin{bmatrix} \boldsymbol{\mu}_X \\ \boldsymbol{\mu}_Y \end{bmatrix} \in \mathbb{R}^{n+m}$, where $\boldsymbol{\mu}_X \in \mathbb{R}^n$ and $\boldsymbol{\mu}_Y \in \mathbb{R}^m$.
* **Covariance Matrix ($\boldsymbol{\Sigma}$):** A symmetric, positive definite matrix $\boldsymbol{\Sigma} \in \mathbb{R}^{(n+m) \times (n+m)}$ partitioned as:
$$\boldsymbol{\Sigma} = \begin{bmatrix} \boldsymbol{\Sigma}_{XX} & \boldsymbol{\Sigma}_{XY} \\ \boldsymbol{\Sigma}_{YX} & \boldsymbol{\Sigma}_{YY} \end{bmatrix}$$
    * $\boldsymbol{\Sigma}_{XX} = \text{Var}(\mathbf{X})$ (size $n \times n$)
    * $\boldsymbol{\Sigma}_{YY} = \text{Var}(\mathbf{Y})$ (size $m \times m$)
    * $\boldsymbol{\Sigma}_{XY} = \text{Cov}(\mathbf{X}, \mathbf{Y}) = \boldsymbol{\Sigma}_{YX}^T$ (size $n \times m$)
* **Determinant ($|\boldsymbol{\Sigma}|$):** The normalization factor ensuring the total integral is 1.

1. Show that the conditional distribution is given by $f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y})$

$$f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) = \frac{1}{\sqrt{(2\pi)^n |\mathbf{S}_{XX}|}} \exp\left( -\frac{1}{2} (\mathbf{x} - \boldsymbol{\mu}^*)^T \boldsymbol{\Sigma}_{X\mid Y}^{-1} (\mathbf{x} - \boldsymbol{\mu}^*) \right),$$
where
* $\boldsymbol{\Sigma}_{X\mid Y} = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}$
* $\boldsymbol{\mu}^* = \boldsymbol{\mu}_x + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_y)$

## Answers for Q2) :
To derive the conditional distribution $f_{\mathbf{X}\mid\mathbf{Y}}(\mathbf{x} \mid \mathbf{y})$, we use the definition of conditional probability density:

$$f_{\mathbf{X}\mid\mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) = \frac{f_{\mathbf{X}, \mathbf{Y}}(\mathbf{x}, \mathbf{y})}{f_{\mathbf{Y}}(\mathbf{y})}$$

Since $\mathbf{Z} = \begin{bmatrix} \mathbf{X} \\ \mathbf{Y} \end{bmatrix}$ is jointly multivariate normal, the marginal distribution of $\mathbf{Y}$ is also multivariate normal, given by $\mathbf{Y} \sim \mathcal{N}(\boldsymbol{\mu}_Y, \boldsymbol{\Sigma}_{YY})$:

$$f_{\mathbf{Y}}(\mathbf{y}) = \frac{1}{\sqrt{(2\pi)^m |\boldsymbol{\Sigma}_{YY}|}} \exp\left( -\frac{1}{2} (\mathbf{y} - \boldsymbol{\mu}_Y)^T \boldsymbol{\Sigma}_{YY}^{-1} (\mathbf{y} - \boldsymbol{\mu}_Y) \right)$$

---

### 1. Matrix Factorization via Schur Complement

To evaluate the ratio of the joint and marginal densities, we use a block-diagonalization of the inverse covariance matrix $\boldsymbol{\Sigma}^{-1}$. Consider the upper block-triangular transformation matrix $\mathbf{A}$:

$$\mathbf{A} = \begin{bmatrix} \mathbf{I}_n & -\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1} \\ \mathbf{0} & \mathbf{I}_m \end{bmatrix}$$

Applying this transformation to $\mathbf{z} - \boldsymbol{\mu}$ yields:

$$\mathbf{A}(\mathbf{z} - \boldsymbol{\mu}) = \begin{bmatrix} \mathbf{I}_n & -\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1} \\ \mathbf{0} & \mathbf{I}_m \end{bmatrix} \begin{bmatrix} \mathbf{x} - \boldsymbol{\mu}_X \\ \mathbf{y} - \boldsymbol{\mu}_Y \end{bmatrix} = \begin{bmatrix} (\mathbf{x} - \boldsymbol{\mu}_X) - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_Y) \\ \mathbf{y} - \boldsymbol{\mu}_Y \end{bmatrix} = \begin{bmatrix} \mathbf{x} - \boldsymbol{\mu}^* \\ \mathbf{y} - \boldsymbol{\mu}_Y \end{bmatrix}$$

where $\boldsymbol{\mu}^* = \boldsymbol{\mu}_X + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_Y)$.

The congruence transformation of the covariance matrix $\boldsymbol{\Sigma}$ using $\mathbf{A}$ diagonalizes the blocks:

$$\mathbf{A} \boldsymbol{\Sigma} \mathbf{A}^T = \begin{bmatrix} \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX} & \mathbf{0} \\ \mathbf{0} & \boldsymbol{\Sigma}_{YY} \end{bmatrix} = \begin{bmatrix} \boldsymbol{\Sigma}_{X\mid Y} & \mathbf{0} \\ \mathbf{0} & \boldsymbol{\Sigma}_{YY} \end{bmatrix}$$

where $\boldsymbol{\Sigma}_{X\mid Y} = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}$ is the Schur complement of $\boldsymbol{\Sigma}_{YY}$ in $\boldsymbol{\Sigma}$.

---

### 2. Determinant Decomposition

Taking the determinant on both sides of the matrix factorization equation, and noting that $|\mathbf{A}| = 1$:

$$|\mathbf{A}| |\boldsymbol{\Sigma}| |\mathbf{A}^T| = \left| \begin{bmatrix} \boldsymbol{\Sigma}_{X\mid Y} & \mathbf{0} \\ \mathbf{0} & \boldsymbol{\Sigma}_{YY} \end{bmatrix} \right| \implies |\boldsymbol{\Sigma}| = |\boldsymbol{\Sigma}_{X\mid Y}| |\boldsymbol{\Sigma}_{YY}|$$

Thus, the normalization factor of the joint density scales as:

$$\frac{1}{\sqrt{(2\pi)^{n+m} |\boldsymbol{\Sigma}|}} = \frac{1}{\sqrt{(2\pi)^n |\boldsymbol{\Sigma}_{X\mid Y}|}} \cdot \frac{1}{\sqrt{(2\pi)^m |\boldsymbol{\Sigma}_{YY}|}}$$

---

### 3. Quadratic Form Decomposition

Inverting both sides of the block-diagonalized covariance equation gives:

$$\boldsymbol{\Sigma}^{-1} = \mathbf{A}^T \begin{bmatrix} \boldsymbol{\Sigma}_{X\mid Y}^{-1} & \mathbf{0} \\ \mathbf{0} & \boldsymbol{\Sigma}_{YY}^{-1} \end{bmatrix} \mathbf{A}$$

Substituting this expression into the quadratic exponent of the joint density function simplifies it to a sum of independent components:

$$(\mathbf{z} - \boldsymbol{\mu})^T \boldsymbol{\Sigma}^{-1} (\mathbf{z} - \boldsymbol{\mu}) = (\mathbf{z} - \boldsymbol{\mu})^T \mathbf{A}^T \begin{bmatrix} \boldsymbol{\Sigma}_{X\mid Y}^{-1} & \mathbf{0} \\ \mathbf{0} & \boldsymbol{\Sigma}_{YY}^{-1} \end{bmatrix} \mathbf{A} (\mathbf{z} - \boldsymbol{\mu})$$

$$= \begin{bmatrix} \mathbf{x} - \boldsymbol{\mu}^* \\ \mathbf{y} - \boldsymbol{\mu}_Y \end{bmatrix}^T \begin{bmatrix} \boldsymbol{\Sigma}_{X\mid Y}^{-1} & \mathbf{0} \\ \mathbf{0} & \boldsymbol{\Sigma}_{YY}^{-1} \end{bmatrix} \begin{bmatrix} \mathbf{x} - \boldsymbol{\mu}^* \\ \mathbf{y} - \boldsymbol{\mu}_Y \end{bmatrix}$$

$$= (\mathbf{x} - \boldsymbol{\mu}^*)^T \boldsymbol{\Sigma}_{X\mid Y}^{-1} (\mathbf{x} - \boldsymbol{\mu}^*) + (\mathbf{y} - \boldsymbol{\mu}_Y)^T \boldsymbol{\Sigma}_{YY}^{-1} (\mathbf{y} - \boldsymbol{\mu}_Y)$$

---

### 4. Synthesizing the Conditional Density

Substituting the split determinant and the decomposed quadratic form back into the joint distribution expression allows the joint density to expand into the following product:

$$f_{\mathbf{X}, \mathbf{Y}}(\mathbf{x}, \mathbf{y}) = \left[ \frac{1}{\sqrt{(2\pi)^n |\boldsymbol{\Sigma}_{X\mid Y}|}} \exp\left( -\frac{1}{2} (\mathbf{x} - \boldsymbol{\mu}^*)^T \boldsymbol{\Sigma}_{X\mid Y}^{-1} (\mathbf{x} - \boldsymbol{\mu}^*) \right) \right] \cdot f_{\mathbf{Y}}(\mathbf{y})$$

Dividing both sides by the marginal density $f_{\mathbf{Y}}(\mathbf{y})$ yields the target conditional density function:

$$f_{\mathbf{X}\mid\mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) = \frac{1}{\sqrt{(2\pi)^n |\boldsymbol{\Sigma}_{X\mid Y}|}} \exp\left( -\frac{1}{2} (\mathbf{x} - \boldsymbol{\mu}^*)^T \boldsymbol{\Sigma}_{X\mid Y}^{-1} (\mathbf{x} - \boldsymbol{\mu}^*) \right)$$

where:

* $\boldsymbol{\Sigma}_{X\mid Y} = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}$
* $\boldsymbol{\mu}^* = \boldsymbol{\mu}_X + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_Y)$

# Q3) Continuous Fault Monitoring of a Spindle Motor

A factory monitors the health of a high-speed CNC spindle. The critical variable is the **Internal Temperature ($X$)**, but because the motor is sealed, they can only observe the **Total Spindle Power ($Y$)**.

**The Setup:**
1.  **Hidden Temperature ($X$):** The standardized temperature deviation follows a standard normal distribution:
    $$X \sim \mathscr{N}(0, 1)$$
2.  **Mechanical Load ($Q$):** The torque required to cut material creates "noise" in the power reading. It is independent of temperature and follows:
    $$Q \sim \mathscr{N}(0, 4)$$
3.  **The Measurement ($Y$):** The power sensor $Y$ measures the combined effect of thermal load and mechanical torque:
$$Y = X + Q$$

---

**Answer the followings:**

**1. Joint Distribution**
Identify the distribution of the random varaible $(X, Y)$.
* Show that $Y \sim \mathscr{N}(0, 5)$.

## 1. Joint Distribution

Because $X$ and $Q$ are independent normal random variables, any linear combination of them remains normally distributed. Therefore, the vector $(X, Y)$ follows a **Bivariate Normal Distribution**.

### Part A: Showing $Y \sim \mathscr{N}(0, 5)$

The observation equation is given by $Y = X + Q$.

* **Mean ($E[Y]$):** By the linearity of expectation:

$$E[Y] = E[X + Q] = E[X] + E[Q] = 0 + 0 = 0$$


* **Variance ($\text{Var}(Y)$):** Since $X$ and $Q$ are independent, their covariance is zero, which allows us to sum their variances directly:

$$\text{Var}(Y) = \text{Var}(X + Q) = \text{Var}(X) + \text{Var}(Q) = 1 + 4 = 5$$



Thus, the marginal distribution of the observation is $Y \sim \mathscr{N}(0, 5)$.

### Part B: Finding the Covariance $\text{Cov}(X, Y)$

Using the bilinear properties of covariance:


$$\text{Cov}(X, Y) = \text{Cov}(X, X + Q) = \text{Cov}(X, X) + \text{Cov}(X, Q)$$

Since $\text{Cov}(X, X) = \text{Var}(X) = 1$, and $\text{Cov}(X, Q) = 0$ due to their independence:


$$\text{Cov}(X, Y) = 1 + 0 = 1$$

## 2. Information Geometry

In measure-theoretic probability, a random variable $X$ is $\sigma(Y)$-measurable if and only if there exists a deterministic Borel-measurable function $f$ such that $X = f(Y)$.

In this system, $X$ is **not** $\sigma(Y)$-measurable. The mapping from the state space $(X, Q)$ to the observation space $Y$ is a many-to-one projection ($Y = X + Q$).

### Practical Interpretation

If the power meter displays $Y = 5$, there are infinitely many combinations of internal temperature ($X$) and mechanical load ($Q$) that satisfy the equation $X + Q = 5$.

The engineer cannot determine the exact value of $X$ because a single scalar measurement cannot uniquely resolve two independent degrees of freedom. Mathematically, it is impossible to distinguish between a dangerously overheated motor cutting a soft material ($X$ is high, $Q$ is low) and a perfectly cool motor cutting a highly dense material ($X$ is low, $Q$ is high).

## 3. Conditional Expectation (The Best Estimate)

For jointly normal random variables, the minimum mean squared error (MMSE) estimator is linear and defined by the conditional expectation formula:


$$E[X \mid Y = y] = \mu_X + \frac{\text{Cov}(X,Y)}{\text{Var}(Y)}(y - \mu_Y)$$

Substituting our derived parameters ($\mu_X = 0$, $\mu_Y = 0$, $\text{Cov}(X,Y) = 1$, $\text{Var}(Y) = 5$):


$$E[X \mid Y = y] = 0 + \frac{1}{5}(y - 0) = \frac{1}{5}y$$

### Signal-to-Total-Variance Ratio Expression

We can express the scaling coefficient $\frac{1}{5}$ structurally using the underlying variances:


$$\frac{\text{Cov}(X,Y)}{\text{Var}(Y)} = \frac{\text{Var}(X)}{\text{Var}(X) + \text{Var}(Q)} = \frac{\sigma_{signal}^2}{\sigma_{signal}^2 + \sigma_{noise}^2}$$

This demonstrates that the optimal estimate scales the observed measurement $y$ by precisely the ratio of the hidden signal's variance to the total observed variance.

## 4. Verification of the Residual

Let the optimal estimator be $\widehat{X} = \frac{1}{5}Y$. The residual error is defined as $Z = X - \widehat{X} = X - \frac{1}{5}Y$.

### Part A: Proving $E[Z] = 0$

Applying the linearity of expectation:


$$E[Z] = E\left[X - \frac{1}{5}Y\right] = E[X] - \frac{1}{5}E[Y] = 0 - \frac{1}{5}(0) = 0$$


The estimation error is zero-mean, confirming that the estimator is globally unbiased.

### Part B: Verifying the Orthogonality Principle $\text{Cov}(Z, Y) = 0$

Evaluating the covariance between the error and the observation:


$$\text{Cov}(Z, Y) = \text{Cov}\left(X - \frac{1}{5}Y, Y\right) = \text{Cov}(X, Y) - \frac{1}{5}\text{Cov}(Y, Y)$$

Knowing that $\text{Cov}(X,Y) = 1$ and $\text{Cov}(Y,Y) = \text{Var}(Y) = 5$:


$$\text{Cov}(Z, Y) = 1 - \frac{1}{5}(5) = 1 - 1 = 0$$

Because $\text{Cov}(Z, Y) = 0$, the residual error is completely uncorrelated with the observation. This satisfies the **Orthogonality Principle**, proving that $\widehat{X}$ has extracted all useful linear information present within $Y$.

## 5. Information Theory: Differential Entropy

The differential entropy of a continuous normal distribution $\mathscr{N}(\mu, \sigma^2)$ is calculated as $h = \frac{1}{2}\ln(2\pi e \sigma^2)$.

### Part A: Computing $h(X)$

Given that $\text{Var}(X) = 1$:


$$h(X) = \frac{1}{2}\ln(2\pi e \cdot 1) = \frac{1}{2}\ln(2\pi e) \text{ nats}$$

### Part B: Computing $h(X \mid Y)$

First, compute the conditional variance $\sigma_{X|Y}^2$:


$$\sigma_{X|Y}^2 = \sigma_X^2 - \frac{\text{Cov}(X,Y)^2}{\text{Var}(Y)} = 1 - \frac{1^2}{5} = 1 - \frac{1}{5} = \frac{4}{5} = 0.8$$

Now, substitute this conditional variance into the differential entropy formula:


$$h(X \mid Y) = \frac{1}{2}\ln\left(2\pi e \cdot \frac{4}{5}\right) = \frac{1}{2}\ln(1.6\pi e) \text{ nats}$$

## 6. Mutual Information

### Part A: Calculating $I(X; Y)$

Mutual information measures the information shared between the two variables:


$$I(X; Y) = h(X) - h(X \mid Y) = \frac{1}{2}\ln(2\pi e) - \frac{1}{2}\ln\left(2\pi e \cdot \frac{4}{5}\right)$$

Using logarithmic properties to simplify:


$$I(X; Y) = \frac{1}{2}\ln\left(\frac{1}{4/5}\right) = \frac{1}{2}\ln\left(\frac{5}{4}\right) \approx 0.1116 \text{ nats}$$

### Part B: Percentage of Uncertainty Removed

The proportion of structural uncertainty removed from the hidden variable by observing $Y$ corresponds to the relative reduction in variance:


$$\text{Uncertainty Reduction} = 1 - \frac{\sigma_{X|Y}^2}{\sigma_X^2} = 1 - \frac{4/5}{1} = \frac{1}{5} = 20\%$$

Observing the power meter removes exactly **20%** of the uncertainty regarding the spindle's internal temperature.

### Part C: Changing Mechanical Noise Variance to $\text{Var}(Q) = 100$

If the torque noise variance increases to $100$:

* **Total Variance:** $\text{Var}(Y) = \text{Var}(X) + \text{Var}(Q) = 1 + 100 = 101$
* **Conditional Variance:** $\sigma_{X|Y}^2 = 1 - \frac{1^2}{101} = \frac{100}{101}$
* **Uncertainty Reduction:** $1 - \frac{100/101}{1} = \frac{1}{101} \approx 0.99\%$

When mechanical noise dominates the system, the observation $Y$ becomes heavily corrupted. The power meter's utility drops dramatically, removing less than $1\%$ of the uncertainty regarding the temperature.

## 7. Diagnostic Interpretation

The factory manager’s claim that $\widehat{X} = Y$ is an optimal estimate is incorrect due to a fundamental misunderstanding of noise corruption.

While $Y$ is globally unbiased ($E[Y] = E[X] = 0$), it is highly **conditionally biased**. If the manager observes an extreme measurement like $Y = 5$ and estimates the temperature to be $\widehat{X} = 5$, they ignore the statistical contributions of the independent variables.

Because the mechanical noise variance ($\text{Var}(Q) = 4$) is four times larger than the temperature variance ($\text{Var}(X) = 1$), any large deviation in $Y$ is statistically much more likely to have been caused by a heavy mechanical load rather than thermal overheating.

The optimal conditional expectation formula, $E[X \mid Y] = \frac{1}{5}Y$, mathematically **shrinks** the raw measurement toward the common mean (zero). This shrinkage acts as a bayesian filter that balances the signal against the noise, protecting the engineer from generating false alarms based on raw, unadjusted sensor data.
